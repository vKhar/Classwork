## Problem 1

Each line in the file,primes.txt represents a prime number below 1,000,000.

Write a function that uses the above file to test if a given positive integer below 1,000,000 is a prime number. If the given numbers exceeds 1,000,000, then utilise a separate function to calculate if it is a prime.

````python
def is_prime_large(n): # gary
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

def is_prime(n):
    if n <= 1000000:
        with open("primes.txt") as f:
            if str(n) in f.read().strip().split('\n'):
                return True
            else:
                return False
    else:
        return is_prime_large(n)
````

## Problem 3

A log of a certain boss fight from a certain game is stored inside the file `log_ex8.csv` under the folder `resources`. The csv file contains the following header.

- `'Parse%'`  refers to the performance of the character relative to all other players in the world,
- `'Name'` refers to the name of the character in the game,
- `'Amount'`  refers to the amount of damage
- `'iLvL'`  refers to the quality of the equipment the character is using,
- `'iLvL%'` refers to the performance of the player relative to players with similar quality of equipment,
- `'Active Time'` refers to the percentage of time spend by the character doing damage during the fight.
- `'DPS'` refers to the amount of damage done per second in the fight.

Your task is to:

- Store the contents of the csv file into a list `records` where each element in the list is itself a list where the elements correspond to the `Parse%`,`Name`,`Amount`,`iLvL`,`iLvL%`,`Active Time`,`DPS` values for each line in the csv file.
- Create an empty list called `player_name`.
- iterate over each element of `records` appending the player names to `player_name`.
- Print each name in `player_name` in alphabetical order.

````python
##Jing Jie
f = open("log_ex8.csv")

records = []
lines = f.readlines() # list of strs
for i in range(len(lines)):
    records.append(lines[i].strip().split(","))
    records[i].pop(-1)
    if len(records[i]) == 8:
        records[i][6] = records[i][6] + "," + records[i][7]
        records[i].pop()
records.pop(0)

player_name = []

for record in records:
    player_name.append(record[1].strip('"'))

alt_player_name = player_name[::]
alt_player_name.sort(reverse = False)
for name in alt_player_name:
    print(name)
````

- store the school name and medal counts in a Python dictionary `medal_count` where the school name is the key and the associated value is a list of the form `[GOLD, SILVER, BRONZE)]`,where GOLD is the number of gold medals, SILVER is the number of silver medals, etc

- print out the top 5 school with the highest number of medals won, regardless of the medal type.

````python
## Lingzhu
f=open('results.txt','r')
lines = f.readlines() #list of lines in file

list_of_line = []
for line in lines:
    line = line.strip('\n')
    list_of_line.append(line)

gold_index = list_of_line.index('Gold')
silver_index = list_of_line.index('Silver')
bronze_index = list_of_line.index('Bronze')
#print(gold_index,silver_index,bronze_index)

medal_count = {}

#getting the schools into the dictionary
for line in list_of_line:
    line = line.strip('\n').split(',') # a line in the form of a list
    school_name = line[-1]
    if school_name in medal_count:
        pass
    else:
        medal_count[school_name] = [0,0,0]
del medal_count['Gold']
del medal_count['Silver']
del medal_count['Bronze']

#getting the n of gold medals
gold_school_name = []
for line in list_of_line[gold_index+1:silver_index]:
    line = line.strip('\n').split(',')
    school_name = line[-1]
    gold_school_name.append(school_name)
    if school_name in medal_count: #no need
        medal_count[school_name][0]= gold_school_name.count(school_name)
                                      #+= 1

#getting the n of silver medals
silver_school_name = []
for line in list_of_line[silver_index+1:bronze_index]:
    line = line.strip('\n').split(',')
    school_name = line[-1]
    silver_school_name.append(school_name)
    if school_name in medal_count:
        medal_count[school_name][1] = silver_school_name.count(school_name)

#getting the n of bronze medals
bronze_school_name = []
for line in list_of_line[bronze_index+1:]:
    line = line.strip('\n').split(',')
    school_name = line[-1]
    bronze_school_name.append(school_name)
    if school_name in medal_count:
        medal_count[school_name][2] = bronze_school_name.count(school_name)


#slight changes to make it a bit neater
medal_count[' Victoria School   '] = medal_count[' Victoria School']
del medal_count[' Victoria School']
medal_count[' Singapore         '] = medal_count[' Singapore']
del medal_count[' Singapore']


print('       School        GOLD  SILVER BRONZE')
for key, value in medal_count.items():
  print(str(key[:19]) + '    ' + str(value[0]) + '     ' + str(value[1]) +'     '+ str(value[2]))

#printing top 5
total_count = {}
for key, value in medal_count.items():
    total_count[key] = value[0] + value[1] + value[2]

sorted_by_val = sorted(total_count.items(), key = lambda v:v[1])
#sorted() -> returns a sorted list of *specified* iterable object
#lambda -> anonymous function, v -> each of the elements(k,v) in dictionary
#v[0] -> sepcify to use the value to sort, where the key is v[0] and value is v[1]
for i in sorted_by_val[-5:]:
    print(i[0])


````
