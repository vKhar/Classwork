```python
### Read Binary Data ###

f = open("mongo_cmd.gif","rb")
gif_array= f.read()

f = open("hammer_bros.jpeg","rb")
jpeg_array= f.read()
print( ":".join( f"{by:08b}" for by in jpeg_array))  # big endian

f = open("NoSQLv2.pptx","rb")
pptx_array= f.read()
```

```python
## save binary data  in MongoDB
## file collection has a schema as follows
## { <file_type} : Binary }

## with file type as a field
import pymongo
col = pymongo.MongoClient("localhost", 27017)["bin"]["file"]
col.insert_one(
    {
     "gif": gif_array 
    }
)
col.insert_one(
    {
        "jpeg": jpeg_array  
    }
)
col.insert_one(
    {
     "pptx": pptx_array 
    }
)
```

##### Read binary data and render image

```python
import pymongo, io
from PIL import Image ## python -m pip install pillow
col = pymongo.MongoClient("localhost", 27017)["bin"]["file"]
doc = col.find_one( {'jpeg': {"$exists":1}}, {'_id':0})
bin_data = doc['jpeg']
image = Image.open(io.BytesIO(bin_data))
image.show()
```
