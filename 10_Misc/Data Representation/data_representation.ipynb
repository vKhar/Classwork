````python
## helper function to print bytes in binary 
## Python does not have a built-in function to print bits
def print_bin(bytes_array: bytes)->str:
    print( ":".join (f"{i:^8d}" for i in range(len(bytes_array) ) ))
    print( ":".join ( f"{b:08b}" for b in bytes_array ) )
````

#### Representing a  float in IEEE 754/ encoding float point numbers
https://www.wikihow.com/Convert-a-Number-from-Decimal-to-IEEE-754-Floating-Point-Representation

```
8.625
1) seperate whole number portion, fraction portion convert to binary
> 00001000.101
2) normalised 
> 1.000101*2^3
3) convert exponent to bias 127, by adding 127,  
> 3+127= 130 =10000010
4)drop the 1, 
> mantissa = 000101
5) sign = 0
6) concatenate sign(1)+exponent(8)+mantissa(23)
```
0<span style="color:red">1000001 0</span>0001010 00000000 00000000


````python
## Convert a floating point number to 32 bit signed 
import struct

value=8.625
by_array= struct.pack('>d', value) #big endian float
print_bin(by_array)
````
