### Python solution to Problem 3

```python
#L=[1,1,2,1,1,3]
L = [1,1,1,1,1,1,1,1]
min1 = float("inf")
min2 = float("inf")
for i in range(len(L)):
    if L[i] < min1:
        min2=min1
        min1 = L[i]
    elif L[i] < min2 and L[i] != min1:
        min2=L[i]
print(min1, end=",") if min1 != float("inf") else print("None",end=",")
print(min2) if min2 != float("inf") else print("None")
```
