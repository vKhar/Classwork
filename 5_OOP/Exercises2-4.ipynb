A program is to be written to implement a to-do list using object-oriented programming (OOP).
The list shows tasks that need to be done.

Each task is given a category and a description.

The base class will be called `ToDo` and is designed as follows:

<center>

| `ToDo`                      |
|------------------------------------------|
|------------------------------------------|
|  `category: STRING`               |
|  `description: STRING`               |
|------------------------------------------|
|  `constructor(c : STRING, d : STRING)`         |
|  `set_category(s : STRING)`          |
|  `set_description(s: STRING)`            |
|  `get_category(): STRING`       |
|  `get_description(): STRING`  |
|  `summary(): STRING` |

</center>

The `summary()` method returns the category and description as a single string.

### Task 1
Write program code to define the class ToDo.  
<div style="text-align: right">[1]</div>



````python
#Task 1
class ToDo:
    def __init__(self,category,description):
        self.__category = str(category)
        self.__description = str(description)
    def set_category(self,category):
        self.__category = str(category)
    def set_description(self,description):
        self.__description = str(description)
    def get_category(self):
        return (self.__category)
    def get_description(self):
        return (self.__description)
    def summary(self):
        return self.__category + " " + self.__description
    def __repr__(self):
        return self.summary()
    def compare_with(self, td):
        if td.__category > self.__category:
            return -1
        elif td.__category == self.__category:
            if td.__description > self.__description:
                return -1
            elif td.__description == self.__description:
                return 0
            else:
                return 1
        else:
            return 1
````

````python
#Task 3
class DatedToDo(Todo):
    def __init__(self, dt, c, d):
        super().__init__(c,d)
        self.__date = dt
    def get_date(self):
        return self.__date
        
    def summary(self):
        return self.__date + "/" + self.get_category() + "/" + self.get_description()
    def __repr__(self):
        return self.summary()

    def compare_with(self, obj): ### self is DatedToDo, obj can be a ToDo object or a DatedToDo object
        if  isinstance(obj, DatedToDo) and self.__date < obj.get_date():
            return -1
        elif isinstance(obj, DatedToDo) and self.__date > obj.get_date():
            return 1
        else:
            return super().compare_with(obj)
        
````

````python
### Implementation  code (Task 1 to Task 3 ) for the class Game
### You should use a incremental approach, ie make sure Task 1 works before Task 2 ..
### tawenn :)
import random

class Game:
    def __init__(self):
        self.__maze = ["XXXXXXXXXX", "X........X", "X.X.X.XX.X", "X.X......X", "X.XXXXXX.X", "X...O....X", "X.X.XX.X.X",
                       "X.X....X.X", "X.XX.XXX.X", "X........X", "XXXXXXXXXX"]
        self.__win = False
        self.__player_pos=[]

    def get_win(self):
        return self.__win

    def get_maze(self):
        for i in range(0, len(self.__maze)):
            new_line = "" ## self.__maz[i] when this is no longer a str
            for g in range(0, len(self.__maze[i])):
                new_line += self.__maze[i][g]
            print(new_line)

    def randomize(self):
        for i in range(0, len(self.__maze)):
            change_list = list(self.__maze[i])
            self.__maze.pop(i)
            self.__maze.insert(i, change_list)

        p_y = random.randint(1, 9)
        p_y_maze = self.__maze[p_y]
        space_list = []
        for i in range(0, len(p_y_maze)):
            if p_y_maze[i] == ".":
                space_list.append(i)
        p_x = random.choice(space_list) ## 
        p_y_maze.pop(p_x)
        p_y_maze.insert(p_x, "P") # self.__maze[py][p_x] = "P"
        # p_y_maze = "".join(p_y_maze) - list to string
        self.__maze.pop(p_y)
        self.__maze.insert(p_y, p_y_maze)

    def player_pos(self):
        for i in range(0, len(self.__maze)):
            for g in range(0, len(self.__maze[i])):
                if self.__maze[i][g] == "O":
                    return g, i

    def movement(self):
        player_input = input("Enter u/d/l/r: ").lower()
        player_pos = self.player_pos()
        player_x = player_pos[0]
        player_y = player_pos[1]
        if player_input == "u":
            if self.__maze[player_y-1][player_x] == "P":
                self.__win = True
            elif self.__maze[player_y-1][player_x] == "X":
                print("Collision detected-try again")
            else:
                self.__maze[player_y].remove(self.__maze[player_y][player_x])
                self.__maze[player_y].insert(player_x, ".")
                self.__maze[player_y-1].pop(player_x)
                self.__maze[player_y-1].insert(player_x, "O")

        if player_input == "d":
            if self.__maze[player_y+1][player_x] == "P":
                self.__win = True
            elif self.__maze[player_y+1][player_x] == "X":
                print("Collision detected-try again")
            else:
                self.__maze[player_y].remove(self.__maze[player_y][player_x])
                self.__maze[player_y].insert(player_x, ".")
                self.__maze[player_y+1].pop(player_x)
                self.__maze[player_y+1].insert(player_x, "O")

        if player_input == "l":
            if self.__maze[player_y][player_x-1] == "P":
                self.__win = True
            elif self.__maze[player_y][player_x-1] == "X":
                print("Collision detected-try again")
            else:
                self.__maze[player_y].remove(self.__maze[player_y][player_x])
                self.__maze[player_y].insert(player_x, ".")
                self.__maze[player_y].pop(player_x-1)
                self.__maze[player_y].insert(player_x-1, "O")

        if player_input == "r":
            if self.__maze[player_y][player_x+1] == "P":
                self.__win = True
            elif self.__maze[player_y][player_x+1] == "X":
                print("Collision detected-try again")
            else:
                self.__maze[player_y].remove(self.__maze[player_y][player_x])
                self.__maze[player_y].insert(player_x, ".")
                self.__maze[player_y].pop(player_x+1)
                self.__maze[player_y].insert(player_x+1, "O")
````
