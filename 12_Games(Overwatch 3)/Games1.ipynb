````python
#YOUR_CODE_HERE : Zheng Yuan
import random
class sudoku:
    def __init__(self):
        self.grids = [['4','3','2','1'],
                      ['1','2','4','3'],
                      ['3','4','1','2'],
                      ['2','1','3','4']
                      ]
    def display(self):
        ret = ""
        #print(len(self.grids))
        for row in self.grids:
            ret += " ".join(row) #converts each row into a string
            ret += "\n" #linebreaks to 'stack' each row on top of each other
        print(ret)
    def transform(self,number):
        #print("1")
        if number == '1':
            rng = random.randint(0,1)
            if rng == 1:
                self.grids[0],self.grids[1] = self.grids[1],self.grids[0] #swap row 0 and 1
                return "1: Swapped 2 top rows in the same quadrants"
            else:
                self.grids[2],self.grids[3] = self.grids[3],self.grids[2] #swap row 2 and 3
                return "1: Swapped 2 bottom rows in the same quadrants"
        elif number == '2':
            #print("2")
            rng = random.randint(0,1)
            if rng == 1:
                for row in self.grids:
                    row[0],row[1] = row[1],row[0] #swap column 0 and 1
                return "2: Swapped 2 leftmost columns in the same quadrants"
            else:
                for row in self.grids:
                    row[2],row[3] = row[3],row[2] #swap column 2 and 3
                return "2: Swapped 2 rightmost columns in the same quadrants"
        elif number == '3':
            #print("3")
            self.grids[0],self.grids[3] = self.grids[3],self.grids[0] #swap row 0 and 3
            return "3: Swapped the top and bottom quadrant rows entirely"
        elif number == '4':
            #print("4")
            for row in self.grids:
                row[0],row[3] = row[3],row[0] #swap column 0 and 3
            return "4: Swapped the left and right quadrant columns entirely"
    def randomTransform(self):
        self.display()
        print(self.transform(str(random.randint(1,4)))) #randomly generate transformation
        self.display()
        print(self.transform(str(random.randint(1,4))))
        self.display()

s = sudoku()
#s.display()
s.randomTransform()


        
````
