````python
#YOUR_CODE_HERE
import random

class Sudoku :
    board = None
    display = None
    tries = 0
    def __init__(self):
        self.board = [ [4,3,2,1],[1,2,4,3],[3,4,1,2],[2,1,3,4] ]
        self.display = [ [1,1,1,1] for _ in range(4)]
    def __repr__(self):
        ret = []
        for row in range(4):
            ret.append("".join([f"{self.board[row][col]:^3d}" if self.display[row][col] else f"{'_':^3}" for col in range(4)])+"\n")
        return "".join( [ line for line in ret] )
    
    def transform1(self):
        print("transform 1")
        if random.randint(0,1):
            # swap 0,1
            self.board[0], self.board[1] = self.board[1], self.board[0]
        else:
            # swap 2,3
            self.board[2], self.board[3] = self.board[3], self.board[2]
    def transform2(self):
        print("transform 2")
        if random.randint(0,1):
            #swap col 0,1
            for row in range(4):
                self.board[row][0], self.board[row][1] = self.board[row][1], self.board[row][0]
        else:
            #swap col 2,3
            for row in range(4):
                self.board[row][2], self.board[row][3] = self.board[row][3], self.board[row][2]
    def transform3(self):
        print("transform 3")
        self.board[0], self.board[2] = self.board[2], self.board[0]
        self.board[1], self.board[3] = self.board[3], self.board[1]   

    def transform4(self):
        print("transform 4")
        for row in range(4):
            self.board[row][0], self.board[row][2] = self.board[row][2], self.board[row][0]

        for row in range(4):
            self.board[row][1], self.board[row][3] = self.board[row][3], self.board[row][1]

    def shuffle(self):
        transforms = [ self.transform1, self.transform2, self.transform3, self.transform4]
        for i in random.sample(range(4),2):
            transforms[i]()

### To make it a game

    def hide(self):
        for row in range(4):
            self.display[row][random.randint(0,3)]=0

    def hide2(self):
        for row in range(4):
            for col in random.sample(range(0,3),2):
                self.display[row][col]=0

    def init_config(self, mode="easy"):
        self.shuffle()
        if mode == "hard":
            self.hide2()
        else:
            self.hide()

    def draw(self):
        print("    "+"".join([f"{i:^3d}" for i in range(1,len(self.board[0])+1)]))
        print("---"+"---"*len(self.board[0]))
        ret=[]
        for row in range(4):
            ret.append(f"{row+1:<3d}|" + "".join([f"{self.board[row][col]:^3d}" if self.display[row][col] else f"{'_':^3}" for col in range(4)])+"\n")
        print("".join(ret))

    def make_move(self):
        self.tries +=1
        while True:
            try:
                row,col,value = map( int,input("Enter row,column,value").split(","))
                assert( 1 <= row <= len(self.board))
                assert( 1 <= col <= len(self.board[0]))
                return (row-1, col-1, value)
            except:
                print("Invalid input")
                continue

    def validate_move(self, move):
        ## This solution simply compares with the value in the initial config
        ## A better solution is to use the rules of sudoku to validate
        row, col, value = move
        if self.display[row][col] == 0 and self.board[row][col] == value:
            return True
        else:
            return False

    def update_config(self, move):
        row, col, value = move
        self.display[row][col] = 1

    def check_win(self):
        if  all ([ all(row) for row in self.display ]):
            return True
        else:
            return False
````

````python
s = Sudoku()
s.init_config()

while True: # game loop
    s.draw()
    move = s.make_move()
    if s.validate_move(move):
        s.update_config(move)
    else:
        print("Invalid Move", move)
        continue
    if s.check_win():
        s.draw()
        print(f"You Win in {s.tries} attempts")
        break
````
