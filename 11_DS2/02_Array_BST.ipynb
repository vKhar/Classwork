<center>

<center>

| Identifier | Data Type | Description |
|-|-|-|
| `ThisTree` | `ARRAY[20] : Node` | The tree data |
| `Root` | `INTEGER` | Index for the root position of the `ThisTree` array |
| `NextFreePosition` | `INTEGER` | Index for the next unused node |

````python
### Code here for BST including AddItemToBinaryTree ,OutputData, Inorder
# Ta Wenn 
class Node:
    def __init__(self, data):
        self.data = data
        self.leftP = -1
        self.rightP = -1
    def __repr__(self):
        return f"{self.data}"

class Bst:
    def __init__(self):
        self.thistree = [Node(None) for i in range(20)]
        for i in range(0, 19):
            self.thistree[i].leftP = i+1
        self.root = -1
        self.nextfreeposition = 0
    def additemtobinarytree(self, newfreeitem):
        if self.nextfreeposition == -1:
            return False
        if self.root == -1:
            self.thistree[self.nextfreeposition].data = newfreeitem
            self.root = self.nextfreeposition
        else:
            cur_position = self.root
            lastmove = "X"
            while cur_position != -1 # and cur_position != self.nextfreeposition:
                prev_position = cur_position
                if newfreeitem < self.thistree[cur_position].data:
                    lastmove = "L"
                    cur_position = self.thistree[cur_position].leftP
                else:
                    lastmove = "R"
                    cur_position = self.thistree[cur_position].rightP
            if lastmove == "R":
                self.thistree[prev_position].rightP = self.nextfreeposition
                self.thistree[self.nextfreeposition].data = newfreeitem
            else:
                self.thistree[prev_position].leftP = self.nextfreeposition
                self.thistree[self.nextfreeposition].data = newfreeitem
        nextfree = self.nextfreeposition
        self.nextfreeposition = self.thistree[self.nextfreeposition].leftP
        self.thistree[nextfree].leftP = -1
        
    def outputdata(self):
        return f"{self.root}, {self.nextfreeposition}, {self.thistree}"
    
    def inorder(self):
        def _inorder(node):
            if node.data == None:
                return
            _inorder(self.thistree[node.leftP])
            print(node)
            _inorder(self.thistree[node.rightP])
        if self.root == -1:
            return
        _inorder(self.thistree[self.root])
            
    
bst = Bst()
i = input("new item: ")
while i != "xxx":
    bst.additemtobinarytree(i)
    print(bst.outputdata())
    i = input("new item: ")
bst.inorder()

````

````python
class Tree:
    def __init__(self, data):
        self.left = None
        self.data = data
        self.right = None
    def get_data(self):
        return self.data
    def get_left(self):
        return self.left
    def get_right(self):
        return self.right
    
    def insert(self, data):
        if data < self.data: ## To look for base case, need to know search left or right
            if self.left == None:
                self.left = Tree(data)
                return ## Base case
            else:
                self.left.insert(data)
        else:
            if self.right == None: 
                self.right = Tree(data)
                return ## Base Case
            else:
                self.right.insert(data)

    def in_order(self):
        if self == None:
            return
        if self.left:
            self.left.in_order()
        print(self.data)
        if self.right:
            self.right.in_order()
    
    def post_order(self):
        if self == None:
            return 
        if self.left:
            self.left.post_order()
        if self.right:
            self.right.post_order()
        print(self.data)
````

````python
import random
random_numbers = [ random.randint(0,999) for _ in range(10) ]

random.sample(range(0,999), 10)
print(random_numbers)

root = Tree(random_numbers[0])


for r in random_numbers[1:]:
    root.insert(r)
root.in_order()
print()
root.post_order()
````
