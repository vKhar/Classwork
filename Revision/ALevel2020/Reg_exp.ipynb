## pre-defined patterns
* \d, \D : digits non-digits
* \w, \W : alphanumeric, non alphanumeric
* \s, \S : whitespaces, non whitespaces

```python
## extract the number of GB in the string
import re
input_str = "I have 800KB of space"
pattern = "(\d+)(KB|MB|GB|TB)"
match = re.search(pattern, input_str)
if match:
    print(match.groups())
else:
    print("invalid")
```

```python
import re
lookup = {
"KB":10**3,
"MB":10**6,
"GB":10**9,
"TB":10**12,
"KiB":2**10,
"MiB":2**20,
"GiB":2**30,
"TiB":2**40
}
#units = "(KB|MB|GB|TB|KiB|MiB|GiB|TiB)"
units = "|".join( lookup.keys())
units = f"({units})"
    
```

```python
import re
pattern = r"(^\d+\.\d+|^\d+)" + units ## include decimals
match = re.search(pattern,"009.0025KB")
if match:
    print(match.groups())
else:
    print("invalid data")
```

```python
def to_bytes(quantity_of_data):
    pattern = "((^\d+\.\d+)|(^\d+))"+units ## include decimals
    match = re.search(pattern, quantity_of_data)
    if match:
        value, unit = match.group(1),match.group(4)
        return float(value) * lookup[unit]
    else:
        return "invalid data"
```

```python
to_bytes("931.51GiB")
```

```python
def convert(quantity_of_data, target_unit):
    ## check target unit
    match = re.search("^"+units, target_unit)
    if not match:
        return "invalid data"
    else:
        target_unit = match.group()

    ## extract quantity of data, and source unit
    #pattern = "^(\d+)"+units
    pattern = "((^\d+\.\d+)|(^\d+))"+units
    match = re.search(pattern,quantity_of_data)
    if match:
        value, unit = match.group(1),match.group(4)
    else:
        return "invalid data"
    return (float(value) * lookup[unit])/lookup[target_unit]
```

```python
convert("1TB","GiB")
```
