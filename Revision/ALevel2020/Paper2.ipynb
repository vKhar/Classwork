```python
### Task 1.1
def task1_1(string_value) :
    h = 0
    for i in range(len(string_value)) :
        val = 33* ord(string_value[i])
        h = (h+val)%1024
    return h

print(task1_1("Hello"))
print(task1_1("Hallo"))
print(task1_1("Hullo"))
```

```python
### Task 2.2
def merge(A,B):
    ret = []
    while len(A) and len(B):
        if A[0] < B[0]:
            ret.append(A.pop(0))
        else:
            ret.append(B.pop(0))
    return ret + A + B
        


def merge_sort(L):
    if len(L) < 2:
        return L
    else:
        mid = len(L)//2
        return merge(merge_sort(L[:mid]),merge_sort(L[mid:]))

def task2_2(list_of_integers):
    return merge_sort(list_of_integers)

task2_2([56,25,4,98,0,18,4,5,7,0]) == [0, 0, 4, 4, 5, 7, 18, 25, 56, 98]
```

```python
### Task 2.3
def task2_3(filename_in,filename_out):
    data = []
    with open(filename_in,"r") as f:
        file = f.readlines()
        for i in file:
            data.append(int(i.strip()))
    f.close()
    data = task2_2(data)

    with open(filename_out,"w") as f:
        for i in data:
            f.write(str(i)+"\n")
        f.close()


task2_3("random.txt","sorted_random.txt")
```

```python
### Task 3.1
units = {"KB":3, "MB":6, "GB":9, "TB":12}
def task3_1(quantity_of_data):
    if quantity_of_data[-2:] in units:
        return int(quantity_of_data[:-2]) * (10 ** units[quantity_of_data[-2:]])
    else:
        return "invalid data"

print(task3_1("8KB"))
print(task3_1("8MB"))
print(task3_1("8GB"))
print(task3_1("8TB"))
```

```python
### Task 3.2
units2 = {"KiB":10, "MiB":20, "GiB":30, "TiB":40}
def task3_2(quantity_of_data):
    if quantity_of_data[-3:] in units2:
        return int(quantity_of_data[:-3]) * (2 ** units2[quantity_of_data[-3:]])
    elif quantity_of_data[-2:] in units:
        return int(quantity_of_data[:-2]) * (10 ** units[quantity_of_data[-2:]])
    else:
        return "invalid data"

print(task3_2("2KiB"))
print(task3_2("2MiB"))
print(task3_2("2GiB"))
print(task3_2("2TiB"))
print(task3_2("2KB"))
print(task3_2("2MB"))
print(task3_2("2GB"))
print(task3_2("2TB"))
```

```python
### Task 3.3
def task3_3(quantity_of_data,target_unit):
    actual = task3_2(quantity_of_data)
    if actual == "invalid data":
        return "invalid data"
    else:
        if target_unit in units:
            return actual / (10 ** units[target_unit])
        elif target_unit in units2:
            return actual / (2 ** units2[target_unit])
        else:
            return "invalid data"

print(task3_3("512MiB", "GiB"))
print(task3_3("72MB", "KB"))
print(task3_3("9MB", "TB"))
```

### Solution using regular expression

```python
import re
lookup = {
"KB":10**3,
"MB":10**6,
"GB":10**9,
"TB":10**12,
"KiB":2**10,
"MiB":2**20,
"GiB":2**30,
"TiB":2**40
}
#units = "(KB|MB|GB|TB|KiB|MiB|GiB|TiB)$" # end with any of KB,MB.. 
units = "|".join( lookup.keys())
units = f"({units})$"
## Task 3.2
def to_bytes(quantity_of_data):
    pattern = "^(\d+)"+ units # start with an integer follow by units
    match = re.search(pattern, quantity_of_data)
    if match:
        value, unit = match.groups()
        return int(value) * lookup[unit]
    else:
        return "invalid data"
## Task 3.3
def convert(quantity_of_data, target_unit):
    ## check target unit
    match = re.search("^"+units, target_unit)
    if not match:
        return "invalid data"
    else:
        target_unit = match.group()

    ## extract quantity of data, and source unit
    pattern = "^(\d+)"+units
    match = re.search(pattern,quantity_of_data)
    if match:
        value, unit = match.groups()
    else:
        return "invalid data"
    return f"{(int(value) * lookup[unit])/lookup[target_unit]} {target_unit}" 
```

```python
print(convert("512MiB", "GiB"))
print(convert("72MB", "KB"))
print(convert("9MB", "TB"))
print(convert("1TB","GiB")) ## my hard disk size reported by Windows
```

```python
import sqlite3, datetime as dt
from flask import Flask
from flask import render_template
from flask import request
from flask import redirect

con = sqlite3.connect(":memory:")
con.execute("""
CREATE TABLE `People` (
	`PersonID`	INTEGER PRIMARY KEY AUTOINCREMENT,
	`FullName`	TEXT,
	`DateOfBirth`	TEXT,
	`ScreenName`	TEXT,
	`IsAdult`	INTEGER
);
"""  
)
with open("people.txt") as f:
    lines = f.readlines()
    for line in lines:
        record = line.strip().split(",")

        ## Screen Name
        screen_name = "".join([c for c in record[0] if c.isalpha()])
        screen_name+=record[1][5:7]+record[1][-2:]
        if record[2] == "Staff":
            screen_name += "Staff"

        #isAdult
        dob = dt.datetime.strptime(record[1],"%Y-%m-%d")
        #naren_birthday = dt.datetime(year=2005, month=11, day=29,hour=4, minute=30, second=30)
        
        today = dt.datetime.now()
        age = (today - dob).days // 365 ## use days instead of year

        con.execute("""
        INSERT INTO People(Fullname,DateOfBirth,ScreenName,IsAdult) 
        VALUES (?,?,?,?)
        """, (record[0],record[1], screen_name, 1 if age > 18 else 0))
    con.commit()

app = Flask(__name__, template_folder='template')
@app.route("/",methods=["GET"])
def root():
    ## Dump DB Contents
    cur = con.execute(
        """ SELECT * FROM People"""
    )
    print(f"{'ID':<4}{'FullName':<15}{'DOB':<11}{'ScreenName':<22}{'IsAdult':<4}")
    data=[]
    for record in cur:
        type = "Staff" if record[3][-5:] == "Staff" else "Person" if record[4] == 1 else "Student"
        print( f"{record[0]:<4}{record[1]:<15}{record[2]:<11}{record[3]:<22}{record[4]:<4}")
        data.append( (record[1],record[3],type))
    ## 
    return render_template("Task4_3_DongZhengyuan.html", jin_data = data)

app.run()
con.close()
```
