Aggreate the scores for each student and print their number of attempts and average score

Eg
```
tom,3,6
janet,2,6
jerry,3,4
alice,1,5

```



````python
def task1_1(string_value) :
    h = 0
    for i in range(len(string_value)) :
        val = 33* ord(string_value[i])
        h = (h+val)%1024
    return h
def task1_2(seed,string_value) :
    return task1_1(f"{seed}{string_value}")
````
