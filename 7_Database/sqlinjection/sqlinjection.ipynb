## SQL Injection:
**Vulnerability**
* SQL String is concatenated with user inputs without checking
* SQL syntax allows -- to be used as comment to ignore the rest of the SQL string
* SQL syntax uses OR 1=1 to effectively set the condition to TRUE no matter what the previous set of criterias were

### Threat vector 1: Confidentiality Compromise 
  you can append '-- to close the string since -- is a comment that will ignore the rest of the condition
  
**enter:**`apple' -- ` 
  
  This will enable you to login as apple without knowing the password

```python
import sqlite3
conn = sqlite3.connect("shopping.db")
user_id = input("Enter ID:")
passwd = input("Enter password")
sql_str = f"SELECT * FROM Login WHERE ID = '{user_id}' AND password = '{passwd}'"
print(sql_str)
result = conn.execute(sql_str).fetchone()
if result == None:
    print("Invalid id or password")
else:
    print("Login successful")
conn.close()
```

***This is the correct way to avoid the sql injection***

```python
import sqlite3
conn = sqlite3.connect("shopping.db")
user_id = input("Enter ID:")
passwd = input("Enter password")
#sql_str = f"SELECT * FROM Login WHERE ID = '{user_id}' AND password = '{passwd}'"
sql_str = "SELECT * FROM Login WHERE ID = ? AND password = ?"
print(sql_str)
result = conn.execute(sql_str,(user_id,passwd)).fetchone()
if result == None:
    print("Invalid id or password")
else:
    print("Login successful")
conn.close()
```
