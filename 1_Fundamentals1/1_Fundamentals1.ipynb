````python
# Ho Jia Le
while True:
    score = input('Please enter your marks here:')
    if score == "": ## check for both presence and termination condition
        break
    else:
        score = int(score)
        if score >69:
            print('A')
        elif score >59:
            print('B')
        elif score >55:
            print('C')
        elif score >49:
            print('D')
        else:
            print('Your grade is a fail.')
````

````python
# Using another approach, with try except:
score = input('Please enter your marks here:')
while score != "":
    try:
        score = int(score)
        if score >69:
            print('A')
        elif score >59:
            print('B')
        elif score >55:
            print('C')
        elif score >49:
            print('D')
        else:
            print('Your grade is a fail.')
    except:
        print("Invalid input")
    score = input('Please enter your marks here:')
````


# Problem 5 Arithmetic Sequence

In mathematics, an **arithmetic sequence** is a sequence of real numbers
$$u_1,u_2,\cdots, u_n$$
such that the difference between two consecutive terms is constant, i.e. $u_{n+1}-u_n=d$ for some constant $d\in \mathbb{R}$, called the **common difference** and $n\geq 1$.

For example, the sequence $2,5,8,11,14,\dots$ is an arithmetic sequence with first term $2$ and common difference $3$.
>   a = 2, d = 3 find the nth term
>
>2, 5, 8, ... n<sup>th<sup>
>
> Term T<sub>n</sub> = a + (n-1)d
  


````python
n = 3
a = 2
d = 3
## sequence should be 2,5,8,11,...
print(f"The {n}th term is {a+(n-1)*d}") ## Math solution
````

````python
## Using a while loop
## Computing solution for seqeunces that do not have a math formula, Eg Collatz seqeunce ,sequence of prime numbers up to nth term
term = a
count= 1
while count < n:
    term = term + d
    count += 1
print(f"The {count}th term is {term}")
````

````python
## Using a for loop
term = a
for i in range(1,n):
    term = term + d
print(f"term {i+1} is {term}") ## Why Wrong?! Use a Trace Table
````

## Compound Interest

Mr. $B$ was offered an investment opportunity with a company where if he was to invest \$1000, he was guaranteed a 100% yearly interest rate, i.e. Mr. $B$ will be receive \$2000 after 1 year, inclusive of his capital.  

A friend of his that work in the company offers him another deal: same capital \$1000, same yearly interest rate of 100% but applied quarterly for a year, meaning every 3 months, the applied interest rate to the current amount is $\frac{100}{4}\%=25\%$. So, after 3rd month, his investment becomes \$1250. After 6th month, it's \$1562.50. After 9th month, it's \$1953.13.
Under this scheme, how much would Mr. $B$'s investment worth after a full year?

Suppose that instead of applying the interest rate of 100% quarterly, a human error caused the interest rate to be applied *hourly* instead, for a full year. How much value will Mr. $B$'s investment worth then?

````python
## Complete this as homework
## Kavish
## the one-liner
print(1000 * ((100+(100/(365*24)))/100)**(365*24))

## the for loop
investment = 1000
for i in range(365*24):
  investment = investment * ((100+(100/(365*24)))/100)

print(f"{round(investment,2)}")
````
