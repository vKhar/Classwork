### (1) Convert a string of base 2 digits,s to denary


```python
def base2_10(s:str)->int:
    ret=0
    for i in range(len(s)):
        ret += int(s[i]) * (2 ** (len(s)-1-i) )
    return ret
```

```python
int("110011",2), base2_10("110011")
```

### (2) Convert a string of base k digits, s to denary

```python
### O(n)
def basek_10(s:str, k)->int: ## addtional digits are a,b,c,d,  ...z", k <= 36
    ret=0
    for i in range(len(s)):
        ret += int(s[i]) * (k ** (len(s)-1-i) ) # problem when s[i] is not a digit
    return ret
```

### (3) Convert a base 10 value, n  to a string base k digits

```python
def base10_pk(n:int, k:int)->str: ## O(log n)
    res = ""
    map = string.digits + string.ascii_uppercase
    while n>0:
        res = map[n%k] + res
        n = n//k
    return res
```
