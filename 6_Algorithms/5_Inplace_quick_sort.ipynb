The following algorithm is an implementation of a quick sort that operates on an array `Scores`. 

This algorithm assumes that the first element of an array is the zeroth element. This means that `Scores[0]` is the first element in the array.

The quick sort  pseudocode is as follows:

>```
>FUNCTION QuickSort(Scores)
>    QuickSortHelper(Scores, 0, LENGTH(Scores) — 1)
>    RETURN Scores
>ENDFUNCTION
>
>FUNCTION QuickSortHelper(Scores, First, Last)
>    IF First < Last
>        THEN 
>            SplitPoint ← Partition(Scores, First, Last)
>            QuickSortHelper(Scores, First, SplitPoint — 1)
>            QuickSortHelper(Scores, SplitPoint + 1, Last)
>    ENDIF
>    RETURN Scores
>ENDFUNCTION
>
>FUNCTION Partition(Scores, First, Last)
>    PivotValue ← Scores[First]
>    LeftMark ← First + 1
>    RightMark ← Last
>    Done ← FALSE
>    WHILE (Done = FALSE)
>        WHILE LeftMark <= RightMark AND Scores[LeftMark] <= PivotValue
>            LeftMark ← LeftMark + 1
>        ENDWHILE
>        WHILE RightMark >= LeftMark AND Scores[RightMark] > PivotValue 
>            RightMark ← RightMark — 1
>        ENDWHILE
>        IF RightMark < LeftMark
>            THEN
>                Done ← TRUE
>        ELSE
>            Temp ← Scores[LeftMark]
>            Scores[LeftMark] ← Scores[RightMark]
>            ScoresIRightMark] ← Temp
>        ENDIF
>    ENDWHILE
>
>    ##<swap Scores[First] with Scores[RightMark]>##
>    
>    RETURN RightMark
>ENDFUNCTION
>```

````python
## Gary
def quick_sort(ls):
    helper(ls, 0, len(ls)-1)
    return ls

def helper(ls, first, last):
    if first < last:
        split_pt = partition(ls, first, last)
        helper(ls, first, split_pt - 1)
        helper(ls, split_pt + 1, last)
    return ls

def partition(ls, first, last):
    left = first + 1
    right = last
    pivot = ls[first]
    done = False
    while done == False:
        while left <= right and ls[left] <= pivot:
            left += 1
        while right >= left and ls[right] > pivot:
            right -= 1
        if right < left:
            done = True
        else:
            ls[left],ls[right] = ls[right],ls[left]
            
    ls[first],ls[right] = ls[right],ls[first]
    return right
````

````python
def l_partition(A,lb, ub):
    pivot = A[lb]
    i = lb + 1 ## values left of i are <= pivot, values at i and above are > pivot
    j = lb + 1 ## iterator
    while j <= ub:
        if A[j] <= pivot:
            # swap with i
            A[i],A[j] = A[j], A[i]
            i += 1
            j +=1
        else:
            j+=1
    ## i-1 is the pivot position
    A[lb],A[i-1] = A[i-1],A[lb]
    return i-1
````
