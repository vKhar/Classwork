```python
#task 1
class Node:
    def __init__(self,data):
        self.data = data #attributes are not private as they need to be accessed by LinkedList class
        self.next = None
    def __repr__(self):
        return f"{self.data}"

class LinkedList:
    def __init__(self):
        self.head = None 
    def insert_front(self,data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
            return
        else:
            new_node.next = self.head
            self.head = new_node
    def insert_back(self,data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
            return
        prev = None
        cur = self.head
        while cur:
            prev = cur
            cur = cur.next
        prev.next = new_node
    def remove_front(self):
        if self.head == None:
            return None
        else:
            res = self.head
            self.head = self.head.next
            return res
    def remove_back(self):
        if self.head == None:
            return None
        elif self.head.next == None:
            res = self.head
            self.head = None
            return res
        else:
            prev = None
            cur = self.head
            try:
                while cur.next:
                    prev = cur
                    cur =  cur.next
                res = prev.next
                prev.next = None
                return res
            except:
                return None
    def __repr__(self):
        if self.head == None:
            return None
        else:
            prev = None
            cur = self.head
            res = []
            while cur:
                prev = cur
                cur = cur.next
                res.append(prev.data)
            return f"{res}"
            
```

```python
#task 2
class Item:
    def __init__(self,stuff,tag):
        self.stuff = stuff
        self.tag = tag #assume that tag created will always be either "RED" or "GREEN"
    def __repr__(self):
        return f"({self.stuff}, {self.tag})"

class Stack:
    def __init__(self):
        self.red = LinkedList()
        self.green = LinkedList() 

    def push(self,item: Item):
        assert( type(item) == Item) ## Only items are allowed
        return self.red.insert_front(item) if item.tag == "RED" else self.green.insert_front(item)

    def pop(self):
        return self.green.remove_front() if self.red.head == None else self.red.remove_front()
     
    def __repr__(self):
        return f"{self.arr}"
        
```

```python
#task 3
class HashTable:
    def __init__(self,size):
        self.arr = [LinkedList() for _ in range(size)]
        self.size = size
    def _hash(self,thing):
        res = 0
        for c in thing:
            res += ord(c)
        return res % self.size
    def insert(self,data):
        index = self._hash(data)
        try:
            self.arr[index].insert_back(data)
            return True
        except:
            return False

    def remove(self,target):
        index = self._hash(target)
        temp = LinkedList()
        x = self.arr[index].remove_front()
        ## to transfer items from LL1 to LL2,  LL2.insert_back( LL1.remove_front() )
        while x is not None:
            if x.data != target:
                temp.insert_back(x.data)
            x = self.arr[index].remove_front()
        else:
            self.arr[index] = temp
   
    def __repr__(self):
        ret = []
        for ll in self.arr:
            if ll.head == None:
                ret.append("[]")
            else:
                ret.append(ll)
        return f"{ret}"
```
