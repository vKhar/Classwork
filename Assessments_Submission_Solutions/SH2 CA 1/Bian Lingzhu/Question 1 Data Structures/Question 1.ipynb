```python
#Task 1
class Node:
    def __init__(self,data):
        self.head = data
        self.next = None
    
    def __repr__(self):
        return f'{self.head}'
    
class LL:
    def __init__(self):
        self.head = None
        
    def insert_front(self,data):
        node = Node(data)
        if self.head == None:
            self.head = node
        else:
            node.next = self.head
            self.head = node
        return
    
    def insert_back(self,data):
        node = Node(data)
        if self.head == None:
            self.head = node
            return
        elif self.head.next == None:
            self.head.next = node
            return
        else:
            cur = self.head
            prev = None
            while cur != None:
                prev = cur
                cur = cur.next
            prev.next = node
            return
            
    def remove_front(self):
        if self.head == None:
            return 0
        else:
            temp = self.head
            self.head = self.head.next
            return temp
        
    def remove_back(self):
        if self.head == None:
            return 0
        else:
            cur = self.head
            prev = None
            while cur.next != None:
                prev = cur
                cur = cur.next
            if prev == None:
                temp = self.head
                self.head = None
                return temp
            else:
                temp = prev.next
                prev.next = None
                return temp
                
    
    def __repr__(self):
        s = ''
        cur = self.head
        while cur != None:
            s = s + f'{cur} > ' 
            cur = cur.next
        return s
    
```

```python
#Task 2
class PriorityNode:
    def __init__(self,data,colour): #assume colour tag is input seperately from data, colour can only be "RED" or "GREEN"
        self.head = data
        self.next = None
        self.colour = colour
    
    def __repr__(self):
        return f'{self.head}'
    

class PriorityStack:
    def __init__(self):
        self.head = None
        self.red = LL()
        self.green = LL()
    
    def push(self,data,colour):
        node = PriorityNode(data,colour)
        if colour == "RED":
            self.red.insert_front(node)
        else:
            self.green.insert_front(node)
    
    def pop(self):
        if self.red.head == None:
            return self.green.remove_front()
        else:
            return self.red.remove_front()
    
    def __repr__(self): #for visualisation
        return f'[{self.red}],[{self.green}]'
```
