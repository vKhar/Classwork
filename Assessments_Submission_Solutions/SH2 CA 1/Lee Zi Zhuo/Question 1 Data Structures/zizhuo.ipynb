```python
#TASK 1
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
    def __repr__(self):
        return f"{self.data}"

class LinkedList:
    def __init__(self):
        self.head = None
    def __repr__(self):
        ret = []
        cur = self.head
        while cur != None:
            ret.append(cur)
            cur = cur.next
        return f"{ret}"
    
    def insert_front(self,value):
        new_node = Node(value)
        if self.head == None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head = new_node
            return 
    def insert_back(self,value):
        new_node = Node(value)
        if self.head == None:
            self.head = new_node
        else:
            prev = None
            cur = self.head
            while cur != None:
                prev = cur
                cur = cur.next
            prev.next = new_node
            return
    def remove_front(self):
        try:
            if self.head.next == None:
                temp = self.head
                self.head = None
                return temp
            else:
                temp = self.head
                self.head = self.head.next
                return temp
        except:
            return None
    def remove_back(self):
        try:
            if self.head.next == None:
                temp = self.head
                self.head = None
                return temp
            else:
                prev = None
                cur = self.head
                while cur.next != None:
                    prev = cur
                    cur = cur.next
                temp = cur
                prev.next = None
                return cur
        except:
            return None

LL = LinkedList()

LL.insert_front("hi")
LL.insert_back("bye")
print(LL)
print(LL.remove_front())
print(LL.remove_back())
print(LL)
```

```python
## Task 2  ## ASSUME COLOR CODE IS A TUPLE e.g. ("RED",value) or ("GREEN",value)
class PriorityStack:
    def __init__(self):
        self.buffer = LinkedList()
    def push(self,value):
        if value[0] == "RED":
            self.buffer.insert_back(value)
        elif value[0] == "GREEN":
            self.buffer.insert_front(value)
    def pop(self):
        ret = self.buffer.remove_back()
            
        return ret
    
        
        
PS = PriorityStack()
PS.push(("RED",'john'))

PS.push(("GREEN",'Gabriel'))
PS.push(("RED",'Beatrice'))
PS.push(("GREEN",'Tamm'))
print(PS.pop())
print(PS.pop())
print(PS.pop())
print(PS.pop())
```

```python
## TASK 3
class HashTable: ## ASSUME NO DUPLICATES
    def __init__(self,size):
        self.buffer = [LinkedList() for _ in range(size)]
        self.size = size
    def hash(self,string):
        total = 0
        for char in string:
            
            value = ord(char)
            
            total += value
        hash_index = total % self.size
        return hash_index
    def insert(self,value):
        try:
            index = self.hash(value)
            self.buffer[index].insert_back(value)
            return True
            
        except:
            return False
    def remove(self,value):
        found = False
        try:
            holder = []
            index = self.hash(value)
            
            
            check = self.buffer[index].remove_front()
            
            if check.data == value:
                
                found = True
                return found
            else:
                holder.append(check.data)
                while check != None: ## will return none if Linked list is empty
                    check = self.buffer[index].remove_front()
                    if check.data == value:
                        found = True
                        break
                    else:
                        holder.append(check.data)
                for value in holder:
                    self.buffer[index].insert_front(value)
                if found == True:
                    return True
                else:
                    return False     
        except:
            
            return False
        

HT = HashTable(10)

print(HT.hash("hello"))  ## hello and h2 have the same hash value
print(HT.hash("H2"))
print(HT.insert("hello"))
print(HT.insert("World"))
print(HT.insert("Today"))
print(HT.insert("H2"))
print(HT.remove("hello"))
print(HT.remove("H2"))
print(HT.remove("H2")) ## second removal false

        
```
