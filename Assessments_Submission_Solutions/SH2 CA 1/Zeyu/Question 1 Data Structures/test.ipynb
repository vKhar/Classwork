````python
##task 1
class node:
    def __init__(self, dat, link):
        self.data=dat
        self.next=link

class linked_l:
    def __init__(self):
        self.head=None
    def insert_front(self, dat):
        if self.head==None:
            self.head=node(dat, None)
        else:
            self.head=node(dat, self.head)
    def insert_back(self, dat):
        if self.head==None:
            self.head=node(dat, None)
        else:
            prev=self.head
            i=self.head.next
            while i!=None:
                prev=i
                i=i.next
            prev.next=node(dat, None)
    def remove_front(self):
        if self.head==None:
            return None
        elif self.head.next==None:
            old=self.head
            self.head=None
            return old
        else:
            old=self.head
            self.head=self.head.next
            return old
    def remove_back(self, dat):
        if self.head==None:
            return None
        elif self.head.next==None:
            old=self.head
            self.head=None
            return old
        else:
            i=self.head.next
            prev=self.head
            while i.next!=None:
                prev=i
                i=i.next
            old=prev.next
            prev.next=None
            return old
    def __repr__(self):
        ret=[]
        ret.append(self.head.data)
        i=self.head.next
        while i!=None:
            ret.append(i.data)
            i=i.next
        return str(ret)
````

````python
##task 2
class priorityQ:
    def __init__(self):
        self.normal=linked_l()
        self.priority=linked_l()
    def push(self, dat, color):
        if color.upper()=="RED":
            self.priority.insert_front(dat)
        elif color.upper()=="GREEN":
            self.normal.insert_front(dat)
        else:
            return None
    def pop(self):
        try:
            result=self.priority.remove_front().data
            return result
        except:
            return self.normal.remove_front().data
````
