```python
## Task 1
## Since the all attributes need to be accessed within other objects anyways, I will not make them private
class Node:
    def __init__(self, data):
        self.data = data
        self.pointer = None

class LinkedList:
    def __init__(self):
        self.root = None
    def insert_front(self, value):
        node = Node(value)
        node.pointer = self.root
        self.root = node
    def insert_back(self, value):
        if self.root is None:
            self.root = Node(value)
        else:
            current = self.root
            while current.pointer is not None:
                current = current.pointer
            current.pointer = Node(value)
    def remove_front(self):
        if self.root is not None:
            temp = self.root
            self.root = self.root.pointer
            return temp.data
        else:
            return "Nothing to remove"
    def remove_back(self):
        if self.root is None:
            return "Nothing to remove"
        elif self.root.pointer is None:
            temp = self.root
            self.root = None
            return temp.data
        else:
            current = self.root
            while current.pointer.pointer is not None:
                current = current.pointer
            temp = current.pointer
            current.pointer = None
            return temp.data
    def __repr__(self):
        arr = []
        current = self.root
        while current:
            arr.append(current.data)
            current = current.pointer
        return f"{arr}"
```

```python
## Task 2
## The PriorityStack.push() method accepts instances of the Item class. An item object is instantiated in this
## manner: Item(data, code)
class Item:
    def __init__(self, data, code):
        self.content = data
        self.code = code
    def __repr__(self):
        return str(self.content)

class PriorityStack:
    def __init__(self):
        self.ll = LinkedList()
    def push(self, value):
        if value.code == "GREEN":
            # self.ll.insert_back(value)
            arr = []
            while True:
                current = self.ll.remove_back()
                if current == "Nothing to remove":
                    self.ll.insert_back(value)
                    for item in arr[::-1]:
                        self.ll.insert_back(item)
                    break
                elif current.code == "GREEN":
                    arr.append(current)
                elif current.code == "RED":
                    self.ll.insert_back(current)
                    self.ll.insert_back(value)
                    for item in arr[::-1]:
                        self.ll.insert_back(item)
                    break
        elif value.code == "RED":
            self.ll.insert_front(value)
    def pop(self):
        return self.ll.remove_front()
    def __repr__(self):
        return str(self.ll)
```
