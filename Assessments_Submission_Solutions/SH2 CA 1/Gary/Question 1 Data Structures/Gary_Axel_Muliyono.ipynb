```python
#task 1
class Node:
    def __init__(self,data):
        self.data = data #attributes are not private as they need to be accessed by LinkedList class
        self.next = None
    def __repr__(self):
        return f"{self.data}"

class LinkedList:
    def __init__(self):
        self.head = None 
    def insert_front(self,data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
            return
        else:
            new_node.next = self.head
            self.head = new_node
    def insert_back(self,data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
            return
        prev = None
        cur = self.head
        while cur:
            prev = cur
            cur = cur.next
        prev.next = new_node
    def remove_front(self):
        if self.head == None:
            return None
        else:
            res = self.head
            self.head = self.head.next
            return res
    def remove_back(self):
        if self.head == None:
            return None
        elif self.head.next == None:
            res = self.head
            self.head = None
            return res
        else:
            prev = None
            cur = self.head
            try:
                while cur.next:
                    prev = cur
                    cur =  cur.next
                res = prev.next
                prev.next = None
                return res
            except:
                return None
    def __repr__(self):
        if self.head == None:
            return None
        else:
            prev = None
            cur = self.head
            res = []
            while cur:
                prev = cur
                cur = cur.next
                res.append(prev.data)
            return f"{res}"
            
```

```python
#task 2
class Item:
    def __init__(self,stuff,tag):
        self.stuff = stuff
        self.tag = tag #assume that tag created will always be either "RED" or "GREEN"
    def __repr__(self):
        return f"{self.stuff}, {self.tag}"

class Stack:
    def __init__(self):
        self.arr = LinkedList() #dynamically sizedd
    def push(self,item):
        #assuming that every item pushed will always be part of the Item class
        if item.tag == "RED":
            self.arr.insert_front(item)
        else:
            self.arr.insert_back(item)
    def pop(self):
        temp = self.arr.remove_front()
#         print(type(temp.data))
        if temp.data.tag == "RED":
#             print("H")
            return temp.data
        else:
            self.arr.insert_front(temp.data)
            new_temp = self.arr.remove_back()
            return new_temp.data
    def __repr__(self):
        return f"{self.arr}"
        
```
