```python
#Task 2.1
class HashTable:
    def __init__(self,n):
        self.arr = [None for _ in range(n)]
    def hasher(self, key):
        sum = 0
        for i in range(0,len(key)):
            sum += ord(key[i]) * (i+1)
        return sum % len(self.arr)
    def insert(self, key, value):
        try:
            index = self.hasher(key)
            if self.arr[index] == None: self.arr[index] = (key,value)
            else:
                old_index = index
                index += 1
                while self.arr[index] != None and index != old_index:
                    index += 1
                    index = index % len(self.arr)
                if index == old_index: return False
                self.arr[index] = (key,value)
            return True
        except:
            return False
    def find(self,key):
        index = self.hasher(key)
        if self.arr[index] == None or self.arr[index][0] != key:
            old_index = index
            index += 1
            while self.arr[index][0] != key and index != old_index:
                index += 1
                index = index % len(self.arr)
            if index == old_index: return None
            return self.arr[index]
        else:
            return self.arr[index]
    def __repr__(self):
        res = "["
        for i in range(len(self.arr)):
            res += f"{i}:{self.arr[i]},"
        res += "]"
        return res
```

```python
#Task 2.3
class Node:
    def __init__(self,value):
        self.value = value
        self.next = -1
    def __repr__(self):
        return f"{self.value}"
class LinkedList:
    def __init__(self):
        self.head = -1
    def insert(self,index):
        new_node = Node(index)
        if self.head == -1:
            self.head = new_node
        elif self.head.next == -1:
            self.head.next = new_node
        else:
            cur = self.head.next
            while cur.next != -1:
                cur = cur.next
            cur.next = new_node
    def traverse(self):
        res = []
        if self.head == -1:
            return res
        elif self.head.next == -1:
            return [self.head]
        else:
            res = [self.head.value]
            cur = self.head.next
            while cur.next != -1:
                res.append(cur.value)
                cur = cur.next
            res.append(cur.value)
            return res
class OrderedHashTable(HashTable):
    #assume keys cannot be deleted

    def __init__(self,n):
        super().__init__(n)
        self.keys_ll = LinkedList()
    def insert(self,key,value):
        x = super().insert(key,value)
        if x == True:
            self.keys_ll.insert(key)
        return x
    def keys(self):
        res = self.keys_ll.traverse()
        return res

```
