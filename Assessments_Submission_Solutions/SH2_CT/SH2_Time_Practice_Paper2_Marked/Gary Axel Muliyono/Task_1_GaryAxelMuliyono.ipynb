```python
#Task 1.1
def hex2str(s):
    final = ""
    for i in range(0,len(s),2):
        tmp = s[i:i+2]
        converted = int(tmp,16)
        tmpchr = chr(converted)
        final = final + tmpchr
    return final

hex2str("4142432C312C31350a")
```
