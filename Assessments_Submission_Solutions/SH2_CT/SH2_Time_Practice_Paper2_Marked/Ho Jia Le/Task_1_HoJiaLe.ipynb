```python
#Task 1.1
def hex2Str(hex):
    hexlist = {'a':10, 'b': 11,'c':12,'d':13,'e':14 ,'f':15,'A':10, 'B': 11,'C':12,'D':13,'E':14 ,'F':15}
    hex = list(hex)
    temp = []
    for i in range(1,len(hex),2):
        temp.append(hex[i-1] + hex[i])
    ret = []
    for i in temp:
        num = 0
        if i[0] in hexlist:
            num += hexlist[i[0]]*16
        else:
            num += int(i[0])*16
        if i[1] in hexlist:
            num += hexlist[i[1]]
        else:
            num += int(i[1])
        ret.append(chr(num))
    return ''.join(ret)
hex2Str('4142432C312C31350a')
```
