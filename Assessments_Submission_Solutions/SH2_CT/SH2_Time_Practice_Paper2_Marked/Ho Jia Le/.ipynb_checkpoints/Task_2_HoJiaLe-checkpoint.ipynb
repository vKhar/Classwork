```python
#Task 2.1
class HashTable:
    def __init__(self, n):
        self.size = n
        self.array = [None for i in range(self.size)]
    def hasher(self, key):
        sum = 0
        for i in range(len(key)):
            sum += (ord(key[i]) * (i+1))
        return sum % self.size
    def insert(self, keyvalue, value):
        key = self.hasher(keyvalue)
        if key > self.size - 1:
            return False
        data = (keyvalue, value)
        count = 0
        while self.array[key] != None:
            count += 1
            key = (key + 1) % self.size
            if count > self.size:
                return False
        self.array[key] = data
        return True
    def find(self, key):
        index = self.hasher(key)
        count = 0
        while count < self.size:
            if self.array[index][0] == key:
                return self.array[index]
            index = (index + 1) % self.size
            count += 1
        return None
            
        return
    def __repr__(self):
        ret = []
        for i in range(len(self.array)):
            if self.array[i] == None:
                ret.append(str(i) + ": None")
            else:
                ret.append(str(i) + ': (' + str(self.array[i][0]) + ',' + self.array[i][1] + ')')
        return '[' + ', '.join(ret) + ']'
        
```

```python
#Task 2.2
HT = HashTable(3)
HT.insert('www.co', 'web')
HT.insert('sch.org','school')
HT.insert('ai.net','sky net')
print(HT)
print(HT.find("www.co"))
print(HT.find("ai.net"))
```

```python
#Task 2.5
class LLnode:
    def __init__(self,data):
        self.data = data
        self.next = None
class OrderedHashTable(HashTable):
    def __init__(self, n):
        self.size = n
        self.head = -1
        self.array = [None for i in range(self.size)]
        self.historylist = []
    def keys(self):
        ret = []
        cur = self.head
        if cur == -1:
            return None
        ret.append(self.array[cur.data][0])
        while cur.next != None:
            cur = cur.next
            ret.append(self.array[cur.data][0])
        return ret
    def insert(self, keyvalue, value):
        key = self.hasher(keyvalue)
        if self.head == -1:
            node = LLnode(key)
            self.head = node
            self.array[key] = (keyvalue, value)
            return True
        if key > self.size - 1:
            return False
        data = (keyvalue, value)
        count = 0
        while self.array[key] != None:
            count += 1
            key = (key + 1) % self.size
            if count > self.size:
                return False
        self.array[key] = data
        cur = self.head
        while cur.next != None:
            cur = cur.next
        node = LLnode(key)
        cur.next = node
        return True 
    def find(self, key):
        index = self.hasher(key)
        count = 0
        while count < self.size:
            if self.array[index] == None:
                return None
            if self.array[index][0] == key:
                return self.array[index]
            index = (index + 1) % self.size
            count += 1
        return None
    def load_cache(self):
        with open("CACHE.TXT") as f:
            content = f.read().strip().split('\n')
        cache = []
        for i in content:
            temp = i.split(',')
            cache.append([temp[0], temp[1]])
        for i in cache:
            if self.insert(i[0],i[1]) == None:
                return "Cache is full."
        return
    def browse(self, url):
        find = self.find(url)
        if find:
            self.historylist.append(find[0])
            return find[1]
        with open("INTERNET.TXT") as f:
            content = f.read().strip().split('\n')
        internetcache = []
        for i in content:
            temp = i.split(',')
            internetcache.append([temp[0], temp[1]])
        found = False
        for i in internetcache:
            if url == i[0]:
                if self.insert(i[0],i[1]) == False:
                    print(i[1])
                    self.historylist.append(i[0]) #Assuming that user still browses website when cache is full
                    return "Cache is full."
                self.historylist.append(i[0])
                return i[1]
                found = True
                break
        if found == False:
            return "404 NOT FOUND"
    def history(self):
        return self.historylist
    def clear_cache(self):
        self.array = [None for i in range(self.size)]
        self.head = -1
        self.historylist = [] #Assuming history is cleared as well
```

```python
#Task 2.5
HT = OrderedHashTable(6)
HT.load_cache()
print(HT.browse("farm.net"))
print(HT)
print(HT.history())


```
