```python
# Task 1.4

from pymongo import MongoClient

def import_data():
    client = MongoClient("127.0.0.1", 27017)
    coll = client["Overwatch"]["Characters"]

    characters = read_characters().values()
    all_items = read_items()
    formatted_characters = []

    for character in characters:
        character_info = character.ret_dict()
        new_items = []
        for item in character_info["items"]:
            new_items.append({"id": item, "name": all_items[item]})
        character_info["items"] = new_items

        formatted_characters.append(character_info)

    coll.insert_many(formatted_characters)
    return

import_data()
```
