```python
#TASK 1.1
import string




def hex2Str(hex_str):
    ret = ""
    my_dict= {}
    map = string.digits + string.ascii_uppercase ##USE ORD() AND CHR()
    for val in map:
        if val.isdigit():
            h_val = map.index(val) + 30
        else:
            h_val = map.index(val) + 31

        my_dict[h_val] = val
    my_dict["0a"] = "\n"
    my_dict["2c"] = ","


    for i in range(0,len(hex_str),2):
        dig_val = hex_str[i] + hex_str[i+1]


        try:
            ret += my_dict[int(dig_val)]
        except:
            ret += my_dict[dig_val]
    return ret

hex2Str('4142432c312c31350a')
```
