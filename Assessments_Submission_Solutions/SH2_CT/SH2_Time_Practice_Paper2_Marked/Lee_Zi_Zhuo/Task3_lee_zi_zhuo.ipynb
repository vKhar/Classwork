```python
##TASK 3.1
def inplace_merge(A,start,mid,end):
    left = start
    right = mid+1
    while left < right and right <= end:
        if A[left] <= A[right]:
            left = left + 1
        else:
            index = right
            tmp = A[right]
            while index > left:
                A[index] = A[index-1]
                index = index -1
            A[index] = tmp
            right = right + 1
    print(A)
A = [2,4,6,8,1,3,5]
inplace_merge(A,0,3,6)
```
