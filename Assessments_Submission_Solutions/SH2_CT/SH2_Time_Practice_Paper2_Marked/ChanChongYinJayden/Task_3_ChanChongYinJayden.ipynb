```python
 Task 3.1
def inplace_merge(A,start,mid,end):
    left = start
    right = mid + 1
    while left < right and right <= end:
        if A[left] <= A[right]:
            left += 1
        else:
            index = right
            tmp = A[right]
            while index > left:
                A[index] = A[index-1]
                index -= 1
            A[index] = tmp
            left = index + 1
            right += 1

A = [2,4,6,8,1,3,5]
inplace_merge(A,0,3,6)
print(A)
```

```python
# Task 3.3
def merge_sort_helper(A,start,end):
    if start >= end:
        return
    merge_sort_helper(A,start,end)

def merge_sort(A):
    start = 0
    end = len(A)-1
    return merge_sort_helper(A,start,end)
```
