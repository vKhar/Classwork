1.  Implement an algorithm that output the pre-order traversal of a BST, given a post order traversal list and a in order travereral list.
```
    def pre_order( in_order_list, post_order_list):
        pass


````python
def pre_order( in_order_list, post_order_list):
    if len(in_order_list) == 1:
        post_order_list.pop(-1)
        return in_order_list
    elif len(in_order_list) == 0:
        return []
    else:
        root = post_order_list.pop(-1)
        mid = in_order_list.index(root)

        right = pre_order(in_order_list[mid+1:], post_order_list) ## need to solve all the right tree first
        left = pre_order(in_order_list[:mid], post_order_list)
        return [root] + left + right ## follows pre order
````


2. Given the following arrays:

```
Data = [11,5, 2, 3, 9, 6, 10,13,14,17, None, None]
Left = [1, 2,-1,-1, 5,-1, -1,-1,-1,-1, -1, -1]
Right =[7, 4, 3,-1, 6,-1,-1, 8, 9, -1, -1, -1]

```
    a.  Implement a insert_BST() Python function as described in Q2,part c,v
    b.  Implement a Python function to return a Python list in in_order traversal of the BST

````python
Data = [11,5, 2, 3, 9, 6, 10,13,14,17, None, None]
Left = [1, 2,-1,-1, 5,-1, -1,-1,-1,-1, -1, -1]
Right =[7, 4, 3,-1, 6,-1,-1, 8, 9, -1, -1, -1]
````

````python
## 2(a)
def get_free():
    for i in range(len(Data)):
        if Data[i] == None:
            return i
    else:
        return None

def insert(current, data):
  index = get_free()
  if get_free() == None :
    print("Full")
    return False
  if data > Data[current] :
    if Right[current] == -1 :
      Data[index] = data
      Right[current] = index
      return True
    else :
      return insert(Right[current], data)
  else :
    if Left[current] == -1 :
      Data[index] = data
      Left[current] = index
      return True
    else :
      return insert(Left[current], data)
````

````python
## 2(a)
insert(0, 7)
````

````python
## 2(b)

def in_order(node, ret):
    ## L N R
    if node == -1:
        return ret
    in_order(Left[node], ret)
    ret.append(Data[node])
    in_order(Right[node], ret)
    return ret
````
