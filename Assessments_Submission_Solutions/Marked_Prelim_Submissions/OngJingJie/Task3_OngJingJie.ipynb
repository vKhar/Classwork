```python
##Task 3.1

import random

def qsort(words):
    if len(words)<2:
        return words
    else:
        left, right = [], []
        rand_index = random.randint(0, len(words)-1)
        words[rand_index], words[0] = words[0], words[rand_index]
        pivot = words[0]
        for word in words[1:]:
            if word[0]<pivot[0]:
                left.append(word)
            else:
                right.append(word)
        return qsort(left) + [pivot] + qsort(right)
```

```python
## Task 3.3

f = open("WORDS.TXT")
words = []
counter = 1
for word in f.readlines():
    words.append((word.strip(),counter))
    counter += 1
f.close()

sorted_partial = []
while words:
    if len(words)>2500:
        to_be_sorted = words[:2500]
        words = words[2500:]
    else:
        to_be_sorted = words
        words = []
    sorted_tmp = qsort(to_be_sorted)
    sorted_partial.append(sorted_tmp)

def merge(list1, list2):
    ret = []
    while list1 and list2:
        if list1[0][0]<list2[0][0]:
            ret.append(list1.pop(0))
        else:
            ret.append(list2.pop(0))
    return ret + list1 + list2

while len(sorted_partial)>1: ##Takes the first two lists in sorted partial and merges them into one sorted list, repeats until only 1 list left
    sorted_partial[0] = merge(sorted_partial[0], sorted_partial[1])
    sorted_partial.pop(1)
sorted_all = sorted_partial[0]
print(sorted_all)
```
