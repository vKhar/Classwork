```python
#Task 2.3
import csv
data = list(csv.reader(open("ARRIVALS.TXT")))[1:]
# print(data)
#direction 1
print("Direction 1")
for item in data:
    if item[1] == '1':
        cur_time = datetime.strptime(item[-1],"%H%M")
        tmp_data = data
        tmp_data.remove(item)
        for j in tmp_data:
            new_time = datetime.strptime(j[-1],"%H%M")
            if abs(int(j[0])-int(item[0]))==1:
                if (new_time - cur_time).seconds // 60 > 2 and new_time > cur_time:
                    
                    print(item[2],j[2],(new_time - cur_time).seconds // 60)
                elif (cur_time - new_time).seconds // 60 > 2 and cur_time > new_time:
                    print(item[2],j[2],(cur_time - new_time).seconds // 60)
                    
#direction 2
print("Direction 2")
for item in data:
    if item[1] == '2':
        cur_time = datetime.strptime(item[-1],"%H%M")
        tmp_data = data
        tmp_data.remove(item)
        for j in tmp_data:
            new_time = datetime.strptime(j[-1],"%H%M")
            if abs(int(j[0])-int(item[0]))==1:
                if (new_time - cur_time).seconds // 60 > 2 and new_time > cur_time:
                    
                    print(item[2],j[2],(new_time - cur_time).seconds // 60)
                elif (cur_time - new_time).seconds // 60 > 2 and cur_time > new_time:
                    print(item[2],j[2],(cur_time - new_time).seconds // 60)
```

```python
#Task 2.3
import csv
data = list(csv.reader(open("ARRIVALS.TXT")))[1:]
# print(data)
max1 = 0
#direction 1
print("Direction 1")
for item in data:
    if item[1] == '1':
        cur_dist = item[-2]
        tmp_data = data
        tmp_data.remove(item)
        for j in tmp_data:
            new_dist = j[-2]
            if abs(int(j[0])-int(item[0]))==1 and abs(float(cur_dist)-float(new_dist)) > max1:
                max1 = abs(float(cur_dist)-float(new_dist))
print('longest distance between two adjacent bus stops is ',round(max1,2), 'km')
max2 = 0
#direction 2
print("Direction 2")
for item in data:
    if item[1] == '2':
        cur_dist = item[-2]
        tmp_data = data
        tmp_data.remove(item)
        for j in tmp_data:
            new_dist = j[-2]
            if abs(int(j[0])-int(item[0]))==1 and abs(float(cur_dist)-float(new_dist)) > max2:
                max2 = abs(float(cur_dist)-float(new_dist))
print('longest distance between two adjacent bus stops is ',round(max2,2), 'km')     
```

```python
#Task 2.5
import csv
data = list(csv.reader(open("ARRIVALS.TXT")))[1:]
dir1_list,dir2_list = [],[]
for item in data:
    new_item = (datetime.strptime(item[4],"%H%M"),int(item[0]),int(item[2]),float(item[3]))
#     print(new_item)
    if item[1] == '1':
        dir1_list.append(new_item)
    else:
        dir2_list.append(new_item)
dir1_list.sort()
dir2_list.sort()
print(len(dir1_list),len(dir2_list))
#given same length,
for i in range(len(dir1_list)):
    mintime = 999999
    mintime_index = -1
    for j in range(len(dir2_list)):
        if abs(dir2_list[j][0] - dir1_list[i][0]).seconds  // 60< mintime:
            mintime = (dir2_list[i][0] - dir1_list[j][0]).seconds // 60
            mintime_index = j
        if (dir1_list[i][0] - dir2_list[j][0]).seconds  // 60< mintime:
            mintime = (dir1_list[i][0] - dir2_list[j][0]).seconds  // 60
            mintime_index = j
    print(mintime,mintime_index,dir1_list[i],dir2_list[mintime_index])
```
