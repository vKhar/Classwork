```python
#Task 3.1
def qsort(words):
    #assume unique words
    if len(words) < 2:
        return words
    left,right = [],[]
    pivot = words[0]
    for item in words[1:]:
        if item[0] < pivot[0]:
            left.append(item)
        else:
            right.append(item)
    return qsort(left) + [pivot] + qsort(right)
```

```python
qsort([('Orange',23),('Apple',121),('Mango',2),('Durian',4)])
```

```python
#Task 3.3
def addsort(lsx):
    ls = lsx
    res = []
    while ls:
        for i,item in enumerate(ls):
            if item == []:
                ls.pop(i)
        if ls == []: return []
        minv = "ZZZZZZZZZZZZZZZZZZZZ"
        minv_index = -1
        for i,item in enumerate(ls):
            if ls[i][0][0] < minv:
                minv = item[i][0]
                minv_index = i
        reck = ls[minv_index].pop(0)
        res.append(reck)
        print(res)
    return res
#this makes res disappear somehow

def newsort():
    with open("WORDS.TXT") as f:
        words = f.readlines()
    for i,item in enumerate(words):
        words[i] = (item.strip(),i)
    words = words[:300]
    sorted_partial = []
    while words:
        if len(words) > 3001:
            extraction = words[:3000]
            words = words[3000:]
        else:
            extraction = words
            words = []
        sorted_tmp = qsort(extraction)
        sorted_partial.append(sorted_tmp)
    return addsort(sorted_partial)
    

print(newsort())
```

```python
#Task 3.4
def bsearch(ls,left,right,word):
    if left > right:
        return None
    mid = (left+right)//2
    if word < ls[mid][0]:
        return bsearch(ls,left,mid-1,word)
    elif word > ls[mid][0]:
        return bsearch(ls,mid+1,right,word)
    else:
        return ls[mid]
def bin_search(sorted_all, word):
    return bsearch(sorted_all,0,len(sorted_all)-1,word)
```

```python
#Task 3.5
def bsearch2(ls,left,right,word,reslist):
    if left > right:
        return None
    mid = (left+right)//2
    if word[0] < ls[mid][0][0]:
        return bsearch2(ls,left,mid-1,word,reslist)
    elif word[0] > ls[mid][0][0]:
        return bsearch2(ls,mid+1,right,word,reslist)
    else:
        if len(ls[mid][0]) >= len(word):
            if word < ls[mid][0][:len(word)]:
                return bsearch2(ls,left,mid-1,word,reslist)
            elif word[0] > ls[mid][0][:len(word)]:
                return bsearch2(ls,mid+1,right,word,reslist)
            else:
                print(word,ls[mid][0][:len(word)])
                x = bsearch2(ls,mid+1,right,word,reslist)
                y = bsearch2(ls,left,mid-1,word,reslist)
                if x is None: x = []
                if y is None: y = []
                return [ls[mid]] + x + y
def prefix_search(sorted_all, prefix):
    return bsearch2(sorted_all,0,len(sorted_all)-1,prefix,[])
```
