```python
#Task 1.1
#assuming all attributes are public
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
    def __repr__(self):
        return f"{self.data},{self.next}"

class LinkedList:
    def __init__(self):
        self.start = None

    def is_empty(self):
        if self.start is None: return True
        return False

    def insert_front(self,data):
        new_node = Node(data)
        if self.is_empty():
            self.start = new_node
            return
        new_node.next = self.start
        self.start = new_node
        return
    def remove_front(self):
        if self.is_empty():
            return None
        ret = self.start
        self.start = self.start.next
        return ret.data
    def __repr__(self):
        res = "["
        if self.is_empty():
            return res + "]"
        cur = self.start
        while cur is not None:
            res = res + f"{cur.data},"
            cur = cur.next
        res = res[:-1] + "]"
        return res
```

```python
#Task 1.2
class Stack:
    def __init__(self):
        self.top = LinkedList()
    def push(self,item):
        self.top.insert_front(item)
    def pop(self):
        return self.top.remove_front()
    def peek(self):
        if self.is_empty():return None
        return self.top.start.data
    def is_empty(self):
        return self.top.is_empty()
```

```python
#Task 1.4
class Queue:
    def __init__(self):
        self.stack = Stack()
        self.stack_size = 0
    def enqueue(self,item):
        self.stack.push(item)
        self.stack_size += 1
    def dequeue(self):
        if self.stack_size == 0:
            return None
        temp_list = []
        for i in range(self.stack_size):
            temp_list.append(self.stack.pop())
        res = temp_list.pop(-1)
        while temp_list:
            self.stack.push(temp_list.pop(-1))
        self.stack_size -= 1
        return res

```
