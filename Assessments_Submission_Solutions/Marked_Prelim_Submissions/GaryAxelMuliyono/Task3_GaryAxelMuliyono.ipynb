````python
#Task 3.1
def qsort(words):
    #assume unique words
    if len(words) < 2:
        return words
    left,right = [],[]
    pivot = words[0]
    for item in words[1:]:
        if item[0] < pivot[0]:
            left.append(item)
        else:
            right.append(item)
    return qsort(left) + [pivot] + qsort(right)
````

````python
#Task 3.4
def bsearch(ls,left,right,word):
    if left > right:
        return None
    mid = (left+right)//2
    if word < ls[mid][0]:
        return bsearch(ls,left,mid-1,word)
    elif word > ls[mid][0]:
        return bsearch(ls,mid+1,right,word)
    else:
        return ls[mid]
def bin_search(sorted_all, word):
    return bsearch(sorted_all,0,len(sorted_all)-1,word)
````
