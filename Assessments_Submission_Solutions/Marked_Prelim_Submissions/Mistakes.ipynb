```python
# task 1.1

class Node:
    def __init__(self, data):
        self.__data = data
        self.__next = None
    
    def get_data(self):
        return self.__data
    
    def set_data(self, newdata):
        self.__data = newdata
        
    def get_next(self):
        return self.__next
    
    def set_next(self, newnext):
        self.__next = newnext
    
    def __repr__(self):
        return f"{self.__data}"
    
class LinkedList:
    def __init__(self):
        self.start = None
    
    def is_empty(self):
        if self.start == None:
            return True
        else:
            return False
    
    def insert_front(self, data):
        new_node = Node(data)
        if self.is_empty():
            self.start = new_node
        else:
            new_node.set_next(self.start) #set the initial first node to be new node's next
            self.start = new_node #set new node at the starting node
        
    def remove_front(self):
        if self.is_empty():
            print("linked list is empty")
        else:
            ret = self.start
            new_start = self.start.get_next()
            self.start = new_start
            return ret
        
    def __repr__(self):
        if self.is_empty():
            return "[]"
        else:
            ret = []
            cur = self.start
            while cur != None:
                ret.append(cur.get_data())
                cur = cur.get_next()
            return f"{ret}"
        
```

```python
# task 1.2

class Stack(LinkedList):
    def __init__(self):
        super().__init__()
        self.top = self.start
        
    def push(self, item): # item is a Node object
        self.insert_front(item.get_data()) #insert_front takes item that is not a Node
        
    def pop(self):
        return self.remove_front()
    
    def peek(self):
        return self.start
    
    def is_empty(self):
        if self.start == None:
            return True
        else:
            return False
    
```
