```python
##task 3.1
def qsort(words):
    if len(words)<2:
        return words
    else:
        pivot=words.pop(0)
        left=[]
        right=[]
        for i in words:
            if i[0]>pivot[0]:
                right.append(i)
            else:
                left.append(i)
        return qsort(left)+[pivot]+qsort(right)
```

```python
##task 3.2
print(qsort([('apple',0),('none',0),('bear',0)])) ##valid
print(qsort([])) ##boundary
print(qsort([('hi', 7) for i in range(3000)]))##invalid
```

```python
##task 3.3
with open('WORDS.txt') as f:
    fs=f.readlines()
    for i in range(len(fs)):
        fs[i]=(fs[i].strip('\n'), i+1)
    sorted_partial=[]
    while fs!=[]:
        if len(fs)>2899:
            sortL=fs[0:2900]
            fs=fs[2900:]
        else:
            sortL=fs
            fs=[]
        sorted_tmp=qsort(sortL)
        sorted_partial.append(sorted_tmp)
    def merger(a, b):
        ret=[]
        while a and b:
            if a[0][0]>b[0][0]:
                ret.append(b.pop(0))
            else:
                ret.append(a.pop(0))
        return ret+a+b
    while len(sorted_partial)>1:
        a=sorted_partial.pop(0)
        b=sorted_partial.pop(0)
        sorted_partial.append(merger(a,b))
    sorted_all=sorted_partial[0]
```
