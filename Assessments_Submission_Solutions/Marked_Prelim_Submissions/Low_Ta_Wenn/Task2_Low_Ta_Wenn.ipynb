```python
# task 2.3
from datetime import datetime

data = []
with open("ARRIVALS.TXT") as f:
    for line in f:
        data.append(line.strip().split(","))
        
data = data[1:] #remove header
first_direction = []
second_direction = []
for item in data:
    if item[1] == "1": #direction 1
        first_direction.append(item)
    else:
        second_direction.append(item)

def qsort(arr):
    if len(arr) < 2:
        return arr
    else:
        pivot = arr[0]
        left = []
        right = []
        for i in range(1, len(arr)):
            if int(pivot[0]) > int(arr[i][0]): # comparing the first element of lists
                left.append(arr[i])
            else:
                right.append(arr[i])
        return qsort(left) + [pivot] + qsort(right)
sorted_first = qsort(first_direction)
sorted_second = qsort(second_direction)
print("Direction 1")

for i in range(1, len(sorted_first)):
    time_diff = datetime.strptime(sorted_first[i][4], "%H%M") - datetime.strptime(sorted_first[i-1][4], "%H%M")
 
    if time_diff.seconds > 2*60:
        print(f"{sorted_first[i][2]} to {sorted_first[i-1][2]}: {time_diff.seconds/60}m")

print("Direction 2")

for i in range(1, len(sorted_second)):
    time_diff = datetime.strptime(sorted_second[i][4], "%H%M") - datetime.strptime(sorted_second[i-1][4], "%H%M")
 
    if time_diff.seconds > 2*60:
        print(f"{sorted_second[i-1][2]} to {sorted_second[i][2]}: {time_diff.seconds/60}m")
```

```python
# task 2.4

data = []
with open("ARRIVALS.TXT") as f:
    for line in f:
        data.append(line.strip().split(","))
        
data = data[1:] #remove header
first_direction = []
second_direction = []
for item in data:
    if item[1] == "1": #direction 1
        first_direction.append(item)
    else:
        second_direction.append(item)

def qsort(arr):
    if len(arr) < 2:
        return arr
    else:
        pivot = arr[0]
        left = []
        right = []
        for i in range(1, len(arr)):
            if int(pivot[0]) > int(arr[i][0]): # comparing the first element of lists
                left.append(arr[i])
            else:
                right.append(arr[i])
        return qsort(left) + [pivot] + qsort(right)
sorted_first = qsort(first_direction)
sorted_second = qsort(second_direction)

first_longest_dist = 0.0
second_longest_dist = 0.0
print("Direction 1")
for i in range(1, len(sorted_first)):
    if float(sorted_first[i][3]) - float(sorted_first[i-1][3]) > first_longest_dist:
        first_longest_dist = float(sorted_first[i][3]) - float(sorted_first[i-1][3]) 
        
first_longest_dist = round(first_longest_dist,2)
print(f"longest distance between two adjancent bus stops is {first_longest_dist}km")

for i in range(len(sorted_second)):
    if float(sorted_second[i][3]) - float(sorted_second[i-1][3]) > second_longest_dist:
        second_longest_dist = float(sorted_second[i][3]) - float(sorted_second[i-1][3])
        
second_longest_dist = round(second_longest_dist,2)
print("Direction 2")
print(f"longest distance between two adjancent bus stops is {second_longest_dist}km")

```
