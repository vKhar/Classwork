```python
# task 1.1
class Node :
    def __init__(self, data):
        self.data = data
        self.next = None
    def __repr__(self):
           return f"{self.data}"
class LinkedList :
    def __init__(self):
        self.start = None
    def is_empty(self) :
        return self.start == None
    def insert_front(self, data) :
        new_node = Node(data)
        if self.is_empty() :
            self.start = new_node
            new_node.next = None
        else :
            new_node.next = self.start
            self.start = new_node
    def remove_front(self):
        if self.is_empty() :
            return None
        ret = self.start
        self.start = ret.next
        return ret
    def __repr__(self):
        if self.is_empty():
            return "[]"
        ret = []
        tmp = self.start
        while tmp :
            ret.append(str(tmp))
            tmp = tmp.next
        ret = ",".join(ret)
        return f"[{ret}]"
```

```python
# test cases
ll = LinkedList()
for i in range(10):
    ll.insert_front(i)

ll.remove_front()
print(ll)
```

- remove_front() should return the item stored and not a Node, you will see later that this seemingly small mistake will cascade into a BIG problem which is difficult to debug.
- in OOP the Node object should be hidden from the user of the Linked List.

```python
# task 1.2
class stack :
    def __init__(self):
        self.top = LinkedList()
    def push(self, item):
        self.top.insert_front(item)
    def pop(self):
        return self.top.remove_front()
    def peek(self):
        if not self.top.is_empty():
            ret = self.top.remove_front() # we need the top element first
            self.top.insert_front(ret.data) # we need to insert it back into the top (we are only peeking)
            return ret
        else :
            return None
    def is_empty(self):
        return self.top.is_empty()
    def __repr__(self):
        return str(self.top)
```
