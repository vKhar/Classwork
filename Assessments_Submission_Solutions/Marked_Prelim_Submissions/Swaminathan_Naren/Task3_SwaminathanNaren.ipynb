```python
# task 3.1
def qsort(words:list):
    if len(words) < 2:
        return words
    left = []
    right = []
    middle = []
    pivot = words[0][0]
    for word in words :
        if word[0] < pivot :
            left.append(word)
        elif word[0] > pivot :
            right.append(word)
        else :
            middle.append(word)
    return qsort(left)+middle+qsort(right)

qsort([("O"), ("A"), ("M")])
```
