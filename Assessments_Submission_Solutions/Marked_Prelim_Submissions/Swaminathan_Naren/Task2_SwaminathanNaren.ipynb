```python
## Task 2.3 Example of pandas style : No iteration, no print
import datetime, pandas
arrivals = pandas.read_csv("ARRIVALS.TXT")
arrivals['ArrivalTime'] = pandas.to_datetime(arrivals['ArrivalTime'],format= '%H%M' )
tmp=arrivals.groupby("Direction").apply(
    lambda frame:
    frame[["ArrivalTime","BusStopCode"]].sort_values("ArrivalTime")
)
tmp["Time Taken"] = tmp["ArrivalTime"].diff()
tmp.loc[ (tmp['Time Taken'].dt.days >=0) & (tmp["Time Taken"].dt.seconds//60 > 2), ["BusStopCode","Time Taken"] ]
```
