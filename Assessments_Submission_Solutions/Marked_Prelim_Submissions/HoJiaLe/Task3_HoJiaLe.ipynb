```python
#Task 3.1
import random
def qsort(L):
    if len(L) < 2:
        return L
    index = random.randint(0, len(L)-1)
    pivot = L.pop(index)
    left = []
    right = []
    for i in L:
        if i[0] > pivot[0]:
            right.append(i)
        else:
            left.append(i)
    return qsort(left) + [pivot] + qsort(right)
qsort([('Orange',23),('Apple',121),('Mango',2),('Durian',4)])
```

```python
#Task 3.4
def bin_search(L, word):
    lb = 0
    ub = len(L) - 1
    while True:
        if lb > ub:
            return None
        mid = (lb + ub) // 2
        if L[mid][0] == word:
            return L[mid]
        if L[mid][0] < word:
           lb = mid + 1
        else:
            ub = mid - 1
print(bin_search(sorted_all, 'outpost'))
```
