```python
#Task 1.1
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None
class linkedlist:
    def __init__(self):
        self.start = None
    def is_empty(self):
        if self.start == None:
            return True
        return False
    def insert_front(self,data):
        newnode = Node(data)
        if self.is_empty():
            self.start = newnode
            return
        newnode.next = self.start
        self.start = newnode
        return
    def remove_front(self):
        if self.is_empty():
            return
        front = self.start
        self.start = front.next
        return self.start.data #return data of node not the node itself
    def __repr__(self):
        if self.is_empty():
            return '[]'
        temp = []
        cur = self.start
        while cur != None:
            temp.append(cur.data)
        return ','.join(temp)
```

```python
#Task 1.2
def Stack:
    def __init__(self):
        self.top = linkedlist()
    def push(self,data):
        return self.top.insert_front(data)
    def pop(self):
        return self.top.remove_front(data)
    def peek(self):
        return self.top.start
    def is_empty(self):
        return self.top.is_empty()
```
