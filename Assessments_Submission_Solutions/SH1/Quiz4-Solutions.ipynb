### Q4

````python
class A():
	def xx(self):
		print("A.xx")
	def yy(self):
		print("A.yy")
	def zz(self): 
		print("A.zz")

class B(A):
	def yy(self):
		super().yy()
	def zz(self):
		super().zz()
		print("B.zz")

class C(B):
	def yy(self):
		print("C.yy")
		super().yy()
	def zz(self):
		super().zz()
		print("C.zz")

aa = A()
bb = B()
cc = C()

````

````python
##(a)
aa.xx(),aa.yy(),aa.zz()
````

````python
##(b)
print( bb.xx())
print(bb.yy())
print(bb.zz())
````

````python
print(cc.xx())
print(cc.yy())
print(cc.zz())
````

### Q6

````python
##(a)
class Client:
    def __init__(self, n, a, t):
        self.__name = n
        self.__address = a
        self.__telephone = t
    ##(c)
    def __repr__(self):
        return f"({self.__name},{self.__address},{self.__telephone})"
````

````python
##(b)
f=open("data.txt","r")
data=[]
for line in f:
    data.append(Client(*line.strip().split(",")))# * upack list into arguments for constructor
##  same as :
##  r = line.strip().split(",")
##  data.append(Client(r[0],r[1],r[2]))
f.close()
````

````python
for i in range(len(data)):
    print(data[i])
````
