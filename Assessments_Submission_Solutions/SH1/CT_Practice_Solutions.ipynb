````python
def task3_1(s):
    map = {
        "KB":3,
        "MB":6,
        "GB":9,
        "TB":12
    }
    try :
        value = int(s[:-2]) 
        unit = s[-2:]
        if unit not in ("KB","MB","GB","TB"):
            raise Exception()
    except:
        print("invalid data")
    else:
        return value * (10** map[unit])
````

````python
def task3_3(s, target):
    accepted = ("KB","MB","GB","TB","KiB","MiB","GiB","TiB")
    map = {
        "KB":10**3,
        "MB":10**6,
        "GB":10**9,
        "TB":10**12,
        "KiB": 2**10,
        "MiB": 2**20,
        "GiB": 2**30,
        "TiB": 2**40
    }
    try :
        if "i" in s:
            value,unit = int(s[:-3]),s[-3:]  
        else:
            value, unit =  int(s[:-2]),s[-2:]
            
        if unit not in accepted or target not in accepted:
            raise Exception()
    except:
        print("invalid data")
        print(value, unit)
    else:
        return (value * (map[unit]) ) / map[target]
    
````

````python
def fill(room):
    num_rows = len(room)
    num_cols = len(room[0])
    flip=False
    for c in range(num_cols):
        for r in range(num_rows):
            if not flip:
                room[r][c] = candidates.pop(0) if len(candidates) else "EMPTY"
            else:
                room[num_rows-1-r][c] = candidates.pop(0) if len(candidates) else "EMPTY"
        flip = not flip
````

````python
print("MP Room")
fill(MPR)
display(MPR)
print()
print("Tut Room1")
fill(TR1)
display(TR1)
print()
print("Tut Room2")
fill(TR2)
display(TR2)
````

````python
#task 1.1
import datetime

class Staff: 
    def __init__(self, Name, date): 
        self.Name = Name
        self.join_date = date
        
    def display_name(self): 
        print("{}_{}".format(self.Name, self.join_date[-4:]))
    
    def length_of_service(self):
        date = self.join_date
        current_year = datetime.datetime.now().year
        current_month = datetime.datetime.now().month
        current_day = datetime.datetime.now().day
        if current_month > int(date[2:4]): 
            length = current_year - int(date[-4:])
        elif current_month < int(date[2:4]):
            length = current_year - int(date[-4:]) - 1
        elif current_day >= int(date[0:2]):
            length = current_year - int(date[-4:])
        elif current_day < int(date[0:2]):
            length = current_year - int(date[-4:]) - 1
        return length

````

````python
#task 1.3
def date_check(s):
    if s.isdigit() == False: 
        return False
    if len(s) != 8:
        return False
    day = int(s[0:2])
    month = int(s[2:4])
    year = int(s[4:])
    day_check = False
    month_check = False
    year_check = False
    if 1970 <= year <= 2021: # valid range of years
        year_check = True
    if 1 <= month <= 12:  # valid range of months
        month_check = True
    if year % 4 == 0 and month == 2:  # leap year Feb
        if 1 <= day <= 29:
            day_check = True
    elif month == 2:  # non-leap year Feb
        if 1 <= day <= 28:
            day_check = True
    elif month in {1,3,5,7,8,10,12}:  # 31-day months (regardless of year)
        if 1 <= day <= 31:
            day_check = True
    else:   # 30-day months (regardless of year)
        if 1 <= day <= 30:
            day_check = True
    return day_check and month_check and year_check

````

````python
def name_check(s): 
    for item in s:
        if (item.isalpha() or item == ' ') == False:
            return False
    lst = s.split(' ')
    for word in lst:
        if word[0].isupper() == False:
            return False
    return True
        
def class_check(s):
    if len(s) != 4:
        return False
    elif (s[0].isdigit() == False) or (s[-1].isdigit() == False):
        return False
    elif (s[1].isupper() == False) or (s[2].isupper() == False):
        return False
    elif (int(s[0]) != 1) and (int(s[0]) != 2):
        return False
    elif (int(s[-1]) > 9) or (int(s[-1]) < 1):
        return False
    return True
      
````
