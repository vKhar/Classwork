#### Question 1: GCD of a list

```python
def gcd(m,n):
    if n == 0:
        return m
    else:
        return gcd(n, m%n)
```

```python
def gcd_l(numbers:list)-> int:
    if len(numbers) == 2:
        return gcd(*numbers) ## * is an unpacking operator, same as gcd(numbers[0],numbers[1])
    else:
        return gcd( numbers[0],gcd_l(numbers[1:]))
```

#### Question 2 find occurances of a digit


```python
# Iterative solution
def findOccurances(n:int,d:str)->tuple:
    total=0
    ret=[]
    for i in range(1, n+1):
        s=str(i)
        if d in s:
            ret.append(s)
        total+= str(i).count(d)
    return total,ret
```

```python
## Recusrive Solution
def r_findOccurances(n,d):
    def _findOccur(pages,d):
        if len(pages) == 1:
            return pages[0].count(d), [] if d not in pages[0] else [pages[0]]
        else:
            count,found =  _findOccur(pages[1:],d)
            if d in pages[0]:
                count += pages[0].count(d)
                found.append(pages[0])
            return count, found
    pages = [ str(i) for i in range(1,n+1)]
    return _findOccur(pages,d)
```
