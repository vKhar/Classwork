```python
#Task 1.1
def generate_digit(string):
    total_sum = 0
    string = str(string)
    for i in string:
        
        
        if (string.index(i)+1) % 2 == 0:
            
            total_sum = int(i) + total_sum
            
            
        else:
            i = int(i)*3
            total_sum = int(i)+ total_sum
            
        
    remainder = total_sum % 10
    result = 10 - remainder

    return result
    
print(generate_digit(9836428))
```
