```python
#Task 1.1

def generate_digit(digits):
    total = 0
    loop_num = 0
    for digit in digits:
        if loop_num%2 == 0:
            total += int(digit)*3
            loop_num+=1
        else:
            total+= int(digit)
            loop_num+=1
    check_digit = 10 - total%10
    return check_digit
```

```python
#Task 1.2

f = open("7DIGITS_CODE.TXT", "r")

GTIN_iden = []
lines = f.readlines()
for line in lines:
    digits = line.strip("\n")
    GTIN_iden.append(digits + str(generate_digit(digits))+ ",")
    
GTIN_iden[-1] = GTIN_iden[-1][0:-1]

f.close()

f = open("GTIN.TXT", "w")

for iden in GTIN_iden:
    f.write(iden)
f.close()
```
