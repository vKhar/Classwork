```python
#Task 1.1
def generate_digit(d): #assuming all inputs are of length 7 digits
    total = 0
    l=[]
    for i in d:
        l.append(i)
    for i in range(len(l)):
        if i%2 != 0:
            total += int(l[i])
        else:
            total += int(l[i])*3
    check_d = 10-total%10
    return check_d


generate_digit('2814307')
        
```
