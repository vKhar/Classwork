```python
f = open("WORDS.txt")
##assume that WORDS.txt contains at least one computing term and number 
try:
    lines = f.readlines()
    max_number = [[0]] ##list with nested lists that will contain max number and corresponding index in list lines
    min_number = [[int(lines[1].strip())]] ##list with nested lists that will contain min number and corresponsing index in list lines
 
    for i in range(1, len(lines)+1, 2):
        lines[i] = int(lines[i].strip())
        
        if lines[i] < min_number[0][0]:
            min_number = [[lines[i], i]]
        elif lines[i] == min_number[0][0]:
            min_number.append([lines[i], i])
        if lines[i]> max_number[0][0]:
            max_number = [[lines[i], i]]
        elif lines[i] == max_number[0][0]:
            max_number.append([lines[i], i])
    print("The words that appeared the most is:")
    for i in range(len(max_number)):
        print(lines[max_number[i][1]-1], end="")
    print("\nThe words that appeared the least is:")
    for i in range(len(min_number)):
        print(lines[min_number[i][1]-1], end="")
    f.close()
                
except:
    print("Error")
    f.close()
```

### Example of a class with context/lifecycle  management

```python
class Foo:
    def __init__(self, name): ## constructor
        self.__name = name
        print(f"{self.__name} is alive")
    def __del__(self): ## destructor
        print(f"{self.__name} is dead")

    def say(self):
        print(f"Hi, I am {self.__name}")

    ## Context Management 
    def __enter__(self): ## Entry
        print(f"{self.__name} is being opened")
        return self
    def __exit__(self,exc_type, exc_value, traceback): # Exit
        print(f"closing .. doing cleanup for {self.__name}")
```

```python
## Without context mangement
f1 = Foo("joe")
f1=""
```

```python
## With context management
with Foo("joe") as j:
    print(type(j))
    j.say()
```
