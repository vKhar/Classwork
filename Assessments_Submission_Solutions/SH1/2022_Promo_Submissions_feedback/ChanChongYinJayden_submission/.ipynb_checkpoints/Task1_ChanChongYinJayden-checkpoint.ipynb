```python
# Task 1.1
class Matrix():
    def __init__(self,m,n):
        self.__buffer = [[None for _ in range(n)] for _ in range(m)]
        self.__row_count = m
        self.__column_count = n  
    def __repr__(self):
        rep = "[\n"
        for row in self.__buffer:
            for e in row:
                rep += f"{e} "
            rep += "\n"
        rep += "]"
        return rep
    def get_value(self,row,column):
        return self.__buffer[row-1][column-1]
    def get_row(self,row):
        return self.__buffer[row-1]
    def get_column(self,column):
        colNum = column-1
        retCol = []
        for row in self.__buffer:
            retCol.append(row[colNum])
        return retCol
    def get_dimension(self):
        return (self.__row_count,self.__column_count)
    
    def set_value(self,column,row,value):
        self.__buffer[row-1][column-1] = value
    def set_row(self,row,vector):
        if len(vector) == self.__column_count:
            invalid = False
            for e in vector:
                if not (type(e) == int or type(e) == float):
                    print(type(e))
                    invalid = True
                    break
            if not invalid:
                self.__buffer[row-1] = vector
            else:
                print("Invalid values in vector")
        else:
            print("Invalid row length")
    def set_column(self,column,vector):
        if len(vector) == self.__row_count:
            invalid = False
            for e in vector:
                if not (type(e) == int or type(e) == float):
                    invalid = True
                    break
            if not invalid:
                count = 0
                for row in self.__buffer:
                    row[column-1] = vector[count]
                    count += 1
            else:
                print("Invalid values in vector")
        else:
            print("Invalid column length")
```

```python
# Task 1.2
# Test case 1
# Should be valid
test1 = Matrix(2,2)
test1.set_row(1,[1,2])
test1.set_row(2,[3,4])
print(test1)

# Test case 2
# Should be valid
test2 = Matrix(1,3)
test2.set_row(1,[5,6,7])
print(test2)


# Test case 3
# Should be valid
test3 = Matrix(3,1)
test3.set_column(1,[9,8,7])
print(test3)

# Test case 4
# Should be valid
test4 = Matrix(0,0)
print(test4)

# Test case 5
# Should be invalid
test5 = Matrix(2,2)
test5.set_row(1,[1,2])
test5.set_row(2,[3])
print(test5)
```
