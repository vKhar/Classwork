````python
def read_file(file_name:str)-> list:
    try : # handle file io runtime error [1m]
        f = None
        f = open(file_name, "r")
        ret = f.readlines() # convert to list of str 
        f.close() ## close [1m]
        return ret
    except:
        print("File Read Error")
        f.close() if f else "" 
        return []
````

````python
lines = read_file("WORDS.TXT") #  file error handling and list returned
if lines: 
    word_count={}
    for i in range(0,len(lines),2): # [1m] read word and count 
        word  = lines[i].strip() #[1m] strip word/count
        count = int(lines[i+1].strip())
        if count in word_count: # [2m] initialise/update
            word_count[count].append(word)
        else:
            word_count[count] = [word]

    max_count = max(word_count.keys()) #[1m]
    min_count= min(word_count.keys()) #[1m]
    print(f"Highest occurance words:{','.join(word_count[max_count])}, Count:{max_count}")
    print(f"Lowest occurance words:{','.join(word_count[min_count])}, Count: {min_count}")
````
