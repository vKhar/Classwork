```python
## 1.1
def generate_digit(s:str)->int:
    total=0
    for i, digit in enumerate(s,1): #2m index of digit and digit
        if i%2 != 0:
            total += int(digit)*3 #1m
        else:
            total += int(digit) #1m
        print(total)
    r = total % 10
    return r if r == 0 else 10-r #1m
```

```python
a = int(input("Enter a: "))
r = int(input("Enter r: "))
m = int(input("Enter m: "))

def re_sum(a, r, m):
    if m == 1:    #base case 
        total = a 
        terms = f"{a} "
        return total, terms #1m
    else:
        item = a * r**(m-1) #1m
        total, terms = re_sum(a, r, m-1) #recursive step 1m
        return total+item, terms +  f"{item} "#1m

total, terms = re_sum(a,r,m)
print(terms)
print(total)
```

### Task 2.3
Advantage of a iterative solution:
- does not make use of the call stack to store intermediate results and therefore it is efficient in using memory resources.

Disadvantage of a iterative solution.
- A problem may be more easily express in a recusrsive form and therefore solve recursively. For example, we can find the sum of a GS up to nth term by simply adding the nth term to the GS up to the (n-1)th term. 
- There are problems that can only be solved recusively without having to use a stack(additional resources). Example, traversal of a nested data structure.

```python
# Task 3.2
def whoWins(players, m):
    if not players:
        return None # 1m 
    current_player_i = 0 # 1m initialisation
    step=m
    while len(players) > 1: ## 1m loop with terminating condition
        player_hit_i = (current_player_i + step) % len(players) #1m for next player with wraparound
        next_player_name = players[ (player_hit_i+1)%len(players)]
        players.pop(player_hit_i) # 1m remove hit player
        current_player_i = players.index(next_player_name) # 1m get index of next player, index is different after hit player is removed

    return players[0]
```

```python
# Task 3.2 Alternative solution
def whoWins(players, m):
    if not players:
        return None # 1m 
    current_player_i = 0 # 1m initialisation
    step=m
    while len(players) > 1: ## 1m loop with terminating condition
        last_player_i = len(players)-1
        player_hit_i = (current_player_i + step) % len(players) #1m for next player with wraparound
        if player_hit_i == last_player_i:
            current_player_i = 0
        else:
            current_player_i = player_hit_i
        players.pop(player_hit_i) # 1m remove hit player
        

    return players[0]
```

```python
# Your code here
class Staff:
    # 1m 
    def __init__(self, name, appt, base_salary):
        self._name = name
        self._appt = appt
        self._base_salary = base_salary
    #1m getters, setters
    def set_name(self, new_name):
        self._name = new_name

    def set_appt(self, new_appt):
        self._appt = new_appt

    def set_base_salary(self, new_salary):
        self._base_salary = new_salary

    def get_name(self):
        return self._name

    def get_appt(self):
        return self._appt

    def get_base_salary(self):
        return self._base_salary
    #1m 
    def calc_annual_salary(self):
        return self._base_salary * 12
    #1m
    def __repr__(self):
        return f"{self._name:20}|{self._appt:20}|{self.calc_annual_salary():12.2f}"
## Task 4.5
    def compare_with(self, other):
        if self.calc_annual_salary() < other.calc_annual_salary():
            return -1
        elif self.calc_annual_salary() > other.calc_annual_salary():
            return 1
        else:
            return 0
## Task 4.2
# 1m for super(), #1m for getter/setter, 2m for overiding calc_annual_salary
class PermStaff(Staff):
    def __init__(self, name, appt, base_salary, bonus_rating=0):
     
        super().__init__(name, appt, base_salary)
        self._bonus_rating = bonus_rating

    def set_bonus_rating(self, new_bonus_rating):
        self._bonus_rating = new_bonus_rating

    def get_bonus_rating(self):
        return self._bonus_rating

    def calc_annual_salary(self):
        return super().calc_annual_salary() + \
            self._bonus_rating * (self.get_base_salary()*12)


class CommissionStaff(Staff):
    def __init__(self, name, appt, base_salary, commission_rate, revenue_generated=0):
        super().__init__(name, appt, base_salary)
        self._commission_rate = commission_rate
        self._revenue_generated = revenue_generated

    def set_commission_rate(self, new_rate):
        self._commission_rate = new_rate

    def get_commission_rate(self):
        return self._commission_rate

    def set_revenue_generated(self, new_value):
        self._revenue_generated = new_value

    def get_revenue_generated(self):
        return self._revenue_generated
    

    def calc_total_commission(self):
        return self._revenue_generated * self._commission_rate

    def calc_annual_salary(self):
        return super().calc_annual_salary() + self.calc_total_commission()
```

```python
## Task 4.3
f = open("STAFF.CSV","r")
staff = []
next(f) # 1m skip header
for line in f:
    line = line.strip().split(",") #1m line processing
    line[-4] = float(line[-4]) 
    ## case for Staff, Perm staff, Comm Staff
    if line[-1]=="" and line[-2] == "" and line[-3] =="":  
        ## Staff
        staff.append(Staff(*line[:3]))       
    elif line[-1] == "" and line[-2] == "":
        ## Perm Staff
        bonus=  float(line[-3])
        staff.append(PermStaff(*line[:3], bonus))
    else:
        revenue = float(line[-1])
        commision = float(line[-2])
        staff.append(CommissionStaff(*line[:3], commision, revenue))
f.close()
```

```python
for s in staff:
    print(s)
```

```python
top_earners = sorted(staff, key = lambda s: s.calc_annual_salary(), reverse=True)
```

```python
## Task 4.5 
top_earners=[staff[0]]
for s in staff[1:]:
    for i in range(len(top_earners)):
        if s.compare_with(top_earners[i]) > -1:
            top_earners.insert(i, s)
            break
    else:
        top_earners.append(s)
```

```python
for s in top_earners:
    print(s)
```

```python
### Task 5
import random
def randomise2lists(A,B):
    size_a = len(A)
    size_b = len(B)
    random_list = []
    random_index_A = []
    random_index_B = []
    for i in range(size_a):
        random_index_A.append( random.randint(0,size_a-1) )
    for i in range(size_a):
        random_index_B.append(random.randint(0,size_b-1))
    for i in range(size_a):
        random_list.append( A[ random_index_A[i] ])
    for i in range(size_b):
        random_list.append( B[ random_index_B[i] ])
    return random_list
```

```python
## 6.3
## 2m
def random_indexes(m):
    rand_list = []
    while len(rand_list) < m:
        i = random.randint(0,m-1)
        if i not in rand_list:
            rand_list.append(i)
    return rand_list
```
