```python
# Kavish, Modified
def whoWins(players, m, position=0):
    if len(players) == 1:
        return players[0]
    else:
        temp_pos = (position + m)%len(players)
        players.pop((position + m)%len(players))
        return whoWins(players, m, temp_pos)
```

```python
# Task 3.2 Jayden 
def whoWins(players,m):
    playersL = players[:]
    lastDead = 0
    if m > 0:
        while True:
            if len(playersL) == 1:
                break
            toKill = lastDead + m
            while toKill > len(playersL)-1:
                toKill -= len(playersL)
            playersL.pop(toKill)
            lastDead = toKill
        return playersL[0]
    else:
        return ""
```

```python
### Accessing private attributes of Base class
class Base:
    def __init__(self,a,b):
        self.__a = a
        self.__b = b
    def get_a(self):
        return self.__a
    def get_b(self):
        return self.__b

class Child(Base):
    def __init__(self,a, b):
        super().__init__(a,b)
    def calc(self):
        #return self.__a + self.__b 
        return self.get_a() + self.get_b()

c = Child(1,3)
c.calc()
```
