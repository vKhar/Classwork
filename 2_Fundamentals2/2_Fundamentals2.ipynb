### Example: Given a list of integers, implement a solution to 
- find the indexes of a given integer in the list, returns None if not found
- remove a given  integer from the list

````python
## Implementation Code:

def find_index(L:list, target:int)->list: ## the data types requirements are for show only :(
    ret=[]
    for i in range(len(L)):
        if L[i] == target:
            ret.append(i)
    return ret if ret else None

def delete(L, target):
    L = L[::]
    while target in L:
        L.remove(target)
    return L


## The 2 functions are independent of each other (loose coupling)
## Each function has only 1 purpose (tight cohesion)
````

Define a function `averages` which takes in a Python List of integers as an argument and returns the mean, median and mode values of the input lists.  ***You are not allow to use any external modules***
```
To calculate the Mean

Add the integers together and divide by the number of integers.
(The sum of values divided by the number of values).

To determine the Median (Not the same as the statistics median)

Arrange the integers in order, find the integer in the middle. if there are even number of integers use either one in the middle. eg 1,2,3,4, either 2 or 3 is the median


To determine the Mode

Count how many times each value occurs; the value that occurs most often is the mode.
(The most frequently occurring value), there may be more than 1 value that has the occurs the most
```


Read more at: https://www.skillsyouneed.com/num/averages.html

**Task 2**
- Generate a List of 20 random numbers from 1 to 10. Use this as input to your function
- Use the function you defined in Task 1 to print the mean, median and mode.


````python
# Code here by Ta Wenn
def mean(list_of_integers):
    total = 0
    no_of_numbers = len(list_of_integers)
    for i in range(len(list_of_integers)):
        total += list_of_integers[i]
    mean_value = total / no_of_numbers
    #print(mean_value)
    return mean_value

def median(list_of_integers):
    list_of_integers.sort()
    if len(list_of_integers) % 2 == 0:
        median_value = [list_of_integers[int(len(list_of_integers) / 2)],
                        list_of_integers[int((len(list_of_integers) / 2) - 1)]]
    else:
        median_value = list_of_integers[int(len(list_of_integers)/2 - 0.5)]
    return median_value

def mode(list_of_integers):
    repeats = 0
    count_dict = {}
    mode_value = []

    for i in range(0, len(list_of_integers)):
        current_value = list_of_integers[i]
        integer_count = list_of_integers.count(current_value)
        count_dict.update({current_value: integer_count})
        if repeats < integer_count:
            repeats = integer_count
    for i in range(0, len(count_dict)):
        if count_dict[list_of_integers[i]] == repeats:
            mode_value.append(list_of_integers[i])

    return list(set(mode_value)) # remove duplicates

    #return mode_value


def averages(integer_list): ## nested function
    return (mean(integer_list),median(integer_list),mode(integer_list))
````

# Recursive Function:A function that calls itself
<ol>
<li> Base Case: The problem size cannot be broken down further, an explicit solution is returned</li>
<li> Recursive Case: The solution can obtained by making use of the result by calling the same function with a smaller problem size [recursive call] </li>
<li> The problem size must eventually reach the base case </li>
</ol>

### Example 1: Find the factorial of a number, n!
- Define as the product of an integer n,  and all the integers below n

````python
## Recursive Factorial function
def r_factorial(n):
    if n == 1:
        return n 
    else: ## recursive case
        return r_factorial(n-1)* n
````

````python
##iterative
def factorial(n):
    ret = 1
    for i in range(1,n+1):
        ret = ret * i
    return ret
````

````python
import math
factorial(3000) == math.factorial(3000)
````

````python
def max_min(*L):
    def _recur(L):
        if len(L) == 1:
            return L[0], L[0]
        else:
            _max, _min = _recur(L[1:])
            if L[0] > _max:
                _max = L[0]
            if L[0] < _min:
                _min = L[0]
            return _max, _min
    L = list(L)
    if len(L) == 0:
        return None
    if len(L) == 1:
        return L[0]
    return _recur(L)
````

````python
#problem5: Naren
def total(n) :
  if len(n) == 1 :
    return n[0]
  else :
    return n[0] + total(n[1:])

def find_mean(List) :
  return total(List)/len(List)

print(find_mean([1,2,3]))
print(find_mean(['a','b','c']))
print(find_mean([]))
````
