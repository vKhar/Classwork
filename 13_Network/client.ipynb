```python
## Send a huge chunk in 1 send
from socket import socket
f = open("HAMLET.txt", "rb")
bytes_stream = f.read() + chr(0).encode()
f.close()
print("Start of File Transfer")
s = socket()
s.connect( ('127.0.0.1',1234)  )
s.sendall(bytes_stream)
print("End of File Transfer")
```

```python
## Send line by line while reading file
from socket import socket
import time 

f = open("HAMLET.txt", "r")
s = socket()
s.connect( ('127.0.0.1',1234)  )
print("Start of File Transfer")
for l,line in enumerate(f):
    s.sendall(line.encode())
    time.sleep(1)
    if l == 30:
        break
    
f.close()
s.sendall(chr(0).encode())
print("End of File Transfer")
```

```python
### Key phrase to terminate Server
s = socket()
s.connect( ('127.0.0.1',1234)  )
s.sendall(b"bye" + chr(0).encode('utf_16'))
s.close()
```

```python
### Guess Client Code
## Naren
from socket import socket
s = socket()
s.connect(('127.0.0.2',3333))
while True :
    server_message = s.recv(4096).decode()
    print(server_message)
    guess = input("Input your guess : ")
    s.send(guess.encode())
    response_to_guess = s.recv(4096).decode()
    if response_to_guess == "Correct" :
      print("You won")
      break
    else :
      print(response_to_guess)
s.close()
```

```python
## Upload Client
## Jayden
from socket import socket

f = open("HAMLET.txt", "r") ## Hard coded
s = socket()
s.connect(('127.0.0.1',1234))

print("Start of File Transfer")
contents = f.read()
s.sendall(contents.encode())
f.close()
s.sendall(chr(0).encode())
print("Sent EOF indicator")

while True: ## wait for acknowledgement
    rec_msg = s.recv(1024)
    if rec_msg == b"OK":
        s.close()
        break
print("End of File Transfer")
```

```python
### HTTP client
import socket

host = "www.google.com"

s = socket.socket()

s.connect((socket.gethostbyname(host),80))
request = f"GET / HTTP/1.1\r\nHost:{host}\r\n\r\n"
s.sendall(request.encode())
response = s.recv(4096)
print(response.decode())
s.close()

```
