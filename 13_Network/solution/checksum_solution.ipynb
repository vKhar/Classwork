```python
def checksum(data:str)->str: ## str of hexadecimal digits
    ## data is in hex format with even number of bytes
    assert(len(data)%2 == 0)
    #Step 1: group into 4 hexadecimal digits and sum them
    total=0
    for i in range(0,len(data),4): ## group in 2 bytes (16 bit)
      total+= int(data[i:i+4],16)
    total_hex = f"{total:x}"

    #Step 2: check for carry on MSB
    if len(total_hex)>4:
        ## add carry 
        check_sum =  f"{int(total_hex[-4:],16) + int(total_hex[-5],16):x}"
    else:
        check_sum = total_hex

    #Step 3: 1's complement and return the hexadecimal value
    checksum = f"{int(check_sum,16):b}"
    comp = "".join(map(lambda x: "0" if x == "1" else "1",checksum))
    return f"{int(comp,2):04x}"
```

```python
def validate(data:str)->str: ## str of hexadecimal digits
    ## data is in hex format with even number of bytes
    assert(len(data)%2 == 0)
    #Step 1: group into 4 hexadecimal digits and sum them
    total=0
    for i in range(0,len(data),4): ## group in 2 bytes (16 bit)
      total+= int(data[i:i+4],16)
    total_hex = f"{total:x}"

    #Step 2: check for carry on MSB
    if len(total_hex)>4:
        ## add carry 
        check_sum =  f"{int(total_hex[-4:],16) + int(total_hex[-5],16):x}"
    else:
        check_sum = total_hex

    return check_sum
```

```python
payload = "450005dc6d6b40008006dc6ec0a80159cf2e"
transmit_data = payload + checksum(payload)
print(transmit_data)
```

```python
def fib(n) :
  x, y = 0,1
  z= None
  ret=[x,y]
  for _ in range(3, n+1):
    z = x + y
    ret.append(z)
    x, y = y,z
  return ret
```
