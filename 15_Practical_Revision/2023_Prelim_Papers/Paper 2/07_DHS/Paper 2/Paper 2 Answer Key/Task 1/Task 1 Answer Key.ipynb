```python
#Task 1.1
def encrypt(plaintext):

    ciphertext = ""
    value_to_add = len(plaintext)

    for i in range(len(plaintext)):  #1 mark for looping through each character

        ascii_value = ord(plaintext[i]) + value_to_add #1 mark for converting character to ASCII number and performing addition

        if ascii_value > 126:
            ascii_value = ascii_value - 95 #1 mark for looping back to 'space' after '~'

        ciphertext = ciphertext + chr(ascii_value) #1 mark for adding the encrypted character to the main ciphertext

        value_to_add = value_to_add - 1 # 1 mark for reducing the value for addition for subsequent letters

    return ciphertext
```

```python
#testing encryption
print(encrypt('urgent!!!!') == ' {olty%$#"') #1 mark for correct output
```

```python
#Task 1.2
def decrypt(ciphertext):

    plaintext = ""
    value_to_subtract = len(ciphertext)

    for i in range(len(ciphertext)):  #1 mark for looping through each character
        cipher_ASCII_value = ord(ciphertext[i])
        plain_ASCII_value = cipher_ASCII_value - value_to_subtract #1 mark for converting character to ASCII number and performing subtraction

        if plain_ASCII_value < 32: #1 mark for looping back to '~' after 'space'
            plain_ASCII_value = plain_ASCII_value + 95

        plaintext = plaintext + chr(plain_ASCII_value) #1 mark for adding the decrypted character to the main plaintext

        value_to_subtract = value_to_subtract - 1 # 1 mark for reducing the value for subtraction for subsequent letters

    return plaintext
```

```python
#testing decryption
print(decrypt(' {olty%$#"') == 'urgent!!!!')  #1 mark for correct output
```
