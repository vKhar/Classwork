```python
#Task 4.1

# Step 1 - import sqlite3
import sqlite3

# Step 2 - create database connection
connection = sqlite3.connect("election.db")

# 1 mark for creating all three tables
# 1 mark for correct fields for all three tables (including correct Primary keys).

# Step 3a - execute SQL query to create tables
connection.execute('''CREATE TABLE 'Candidate' (
                    'CID' INTEGER PRIMARY KEY,
                    'Name' TEXT NOT NULL,
                    'Gender' TEXT NOT NULL,
                    'Class' TEXT NOT NULL);''')

connection.execute('''CREATE TABLE 'Voter' (
                    'VID' INTEGER PRIMARY KEY,
                    'Name' TEXT NOT NULL,
                    'Gender' TEXT NOT NULL,
                    'Class' TEXT NOT NULL);''')

#1 mark for correct foreign keys
connection.execute('''CREATE TABLE 'Vote' (
                    'VID' INTEGER,
                    'CID' INTEGER,
                    FOREIGN KEY('CID') REFERENCES 'Candidate'('CID'),
                    FOREIGN KEY('VID') REFERENCES 'Voter'('VID'),
                    PRIMARY KEY('VID','CID'));''')

# Step 4 - make changes permanent
connection.commit()

# Step 5 - destroy database connection
connection.close()
```
