```python
# 2.1
class Array:
    def __init__(self, n):
        self.capacity = n
        self.buffer = [ None for _ in range(n)]
    def _validate_index(self,i):
        if  not (-1 < i < len(self.buffer)):
            raise Exception(f"array size is {self.capacity}")
    def add_data(self, new_data, i):
        self._validate_index(i)
        self.buffer[i] = new_data
    def get_data(self, i):
        self._validate_index(i)
        return self.buffer[i]
    def delete_data(self, i):
        self._validate_index(i)
        self.buffer[i] = None
```

```python
class CircularQueue(Array):
    # The inheritanc relationship is WRONG!
    # A CircularQueue is NOT an Array, A CircularQueue has an Array
    # the methods inherited SHOULD NOT BE used on a CircularQueue
    # for answering this question, the code will follow as per required by the question
    def __init__(self, n):
        super().__init__(n+1) ## 1 empty slot is used for queue full detection
        self.head = 0
        self.tail = 0
    def get_head(self):
        return self.head
    def is_empty(self):
        return self.head == self.tail
    def is_full(self):
        return (self.tail + 1)% self.capacity == self.head
    def enqueue(self, data):
        if not self.is_full():
            self.add_data(data, self.tail)
            self.tail = (self.tail+1)%self.capacity
    def dequeue(self):
        if self.is_empty():
            return 
        ret = self.get_data(self.head)
        self.delete_data(self.head)
        self.head = (self.head+1)%self.capacity
        return ret
            
```

```python
#2.2
class Node:
    def __init__(self, data=None):
        self.data = data
        self.left = None
        self.right = None
    def get_data(self):
        return self.data
    def get_left_child(self):
        return self.left
    def get_right_child(self):
        return self.right
    def make_left_child(self, l_child):
        self.left = l_child
    def make_right_child(self, r_child):
        self.right = r_child
```

```python
Root = Node("the")
Node1 = Node("be")
Node2 = Node("to")
Node3 = Node("of")
Node4 = Node("and")
Node5 = Node("a")
Node6 = Node("in")
Node7 = Node("that")
Node8 = Node("have")
Node9 = Node("it")
Node10 = Node("for")
Node11 = Node("not")
Node12 = Node("on")
Node13 = Node("with")
Node14 = Node("as")

Root.make_left_child(Node1)
Node1.make_left_child(Node2)
Node2.make_left_child(Node3)
Node2.make_right_child(Node4)
Node1.make_right_child(Node5)
Node5.make_left_child(Node6)
Node5.make_right_child(Node7)
Root.make_right_child(Node8)
Node8.make_left_child(Node9)
Node9.make_left_child(Node10)
Node9.make_right_child(Node11)
Node8.make_right_child(Node12)
Node12.make_left_child(Node13)
Node12.make_right_child(Node14)
```

```python
## Breadth-First Search
target = input("Search for")
q = CircularQueue(50)
q.enqueue(Root)
while not q.is_empty():
    node = q.dequeue()
    if node.get_data() == target:
        print("Found")
        break
    else:
        left_child, right_child = node.get_left_child(), node.get_right_child()
        if left_child :
            q.enqueue(left_child)
        if right_child:
            q.enqueue(right_child)
else:
    print("Not Found")
```
