```python
# Task 2.1

class Array: # zero-index array
    def __init__(self, n):
        self.container = [] # 1 mark
        for i in range(n): # 1 mark
            self.container.append(None)
        self.capacity = n # 1 mark
    
    def add_data(self, new_data, i): # 1 mark
        self.container[i] = new_data
        
    def get_data(self, i): # 1 mark
        return self.container[i]
    
    def delete_data(self,i): # 1 mark
        self.container[i] = None

# Array: 6 marks

class Circ_Queue(Array): # 1 mark
    def __init__(self, n):
        self.container = []
        for i in range(n):
            self.container.append(None)
            
        self.capacity = n
        self.head = 0 # 1 mark
        self.tail = 0 # 1 mark
    
    def enqueue(self, new_data):
        if (self.tail + 1) % self.capacity == self.head:
            print("Error: Queue is full") # 1 mark
        else:
            self.add_data(self.tail, new_data) # 1 mark
            self.tail = (self.tail + 1) % self.capacity # 1 mark
    
    def dequeue(self):
        if self.head == self.tail: # 1 mark
            print("Error: Queue is empty")
        else:
            output = get_data(self.head) # 1 mark
            self.delete_data(self.head) # 1 mark
            self.head = (self.head + 1) % self.capacity # 1 mark
            return output
    
    def get_head(self):
        return self.get_data(self.head) # 1 mark

# CircQueue: 11 marks
```

```python
# Task 2.2

class Node:
    def __init__(self,data):
        self.data = data # 1 mark
        self.left = None # 1 mark
        self.right = None
    
    def make_left_child(self,lchild): # 1 mark
        self.left = lchild
    
    def make_right_child(self,rchild):
        self.right = rchild
    
    def get_data(self): # 1 mark
        return self.data
    
    def get_left_child(self):
        return self.left
    
    def get_right_child(self):
        return self.right

# Node: 4 marks
```
