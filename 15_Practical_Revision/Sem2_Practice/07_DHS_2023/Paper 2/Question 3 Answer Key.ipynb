```python
#Task 3.1

class Student:
    
    def __init__(self, studentid, full_name, gender, date_of_birth, subject_combination): #2 marks for correct __init__ with correct parameters & list of parameters
        self.__studentid = studentid # assignment of value of parameters to private attributes (all correct - 2 marks, at least 2 attributes correct - 1 mark)
        self.__full_name = full_name
        self.__gender = gender
        self.__date_of_birth = date_of_birth
        self.__subject_combination = subject_combination
        
    def get_studentid(self): #1 mark for all getters
        return self.__studentid
    
    def set_studentid(self, studentid): #1 mark for all setters
        self.__studentid = studentid
    
    
    
    
    def get_full_name(self):
        return self.__full_name
    
    def get_gender(self):
        return self.__gender
    
    def get_date_of_birth(self):
        return self.__date_of_birth
    
    def get_subject_combination(self):
        return self.__subject_combination
    
    
    
    def set_full_name(self, full_name):
        self.__full_name = full_name
        
    def set_gender(self, gender):
        self.__gender = gender
    
    def set_date_of_birth(self, date_of_birth):
        self.__date_of_birth = date_of_birth
    
    def set_subject_combination(self, subject_combination):
        self.__subject_combination = subject_combination
    
    def show_details(self):
        print("Student ID:" , self.__studentid) #1 mark for displaying all attributes correctly (except Subject Combination)
        print("Full Name:" , self.__full_name)
        print("Gender:" , self.__gender)
        print("Date of Birth:" , self.__date_of_birth)
        print("Subject_Combination:")
        for i in range(len(self.__subject_combination)): # 1 mark for displaying Subject Combination correctly
            print(f"Subject {i+1}: {self.__subject_combination[i]}")
```

```python
#Task 3.2

import csv

students = []
with open("STUDENTS.csv", "r", newline="") as infile: #1 mark for opening and closing file/using with & using csv reader or other methods of splitting the data
    records = csv.reader(infile, delimiter = ",")
    for record in records: #1 mark for loop for extracting data     
        #to append all subjects into a list
        subjects = []
        for i in range(len(record[4:])): #1 mark for creating list of subjects
            subjects.append(record[i+4])
            
        #creating new student instance
        newStudent = Student(int(record[0]),record[1],record[2],record[3],subjects) #1 mark for creating Student object
        students.append(newStudent) #1 mark for inserting Student object into students list


#print(len(students))
#print("Students:")
#print()
for student in students: #1 mark for displaying all students' details using show_details method
    student.show_details()
    print()
```

```python
#Task 3.3

def binary_search_recursive(students, search_studentid, first, last):
    if first > last: # 1 mark for Base case for recursion: The recursion will stop when the index of the first item is greater than the index of last
        return -1
    else: # Recursively call the function
        midpoint = (first + last) // 2 # 1 mark for calculating the midpoint

        if search_studentid == students[midpoint].get_studentid(): #1 mark for checking if value of midpoint is correct followed by returning the position
            return midpoint
        elif search_studentid > students[midpoint].get_studentid():  # 1 mark for correct recursive call with correct parameters if the search item is more than the item at the midpoint
            first = midpoint + 1
            #print(f"In >: first = {first}")
            return binary_search_recursive(students, search_studentid, first, last) # Focus on the items after the midpoint
        else: # 1 mark for correct recursive call with correct parameters if the search item is less than the item at the midpoint
            last = midpoint - 1
            #print(f"In else: last = {last}")
            return binary_search_recursive(students, search_studentid, first, last) # Focus on the items before the midpoint

search = "Yes"
while search == "Yes": #1 mark for looping as long as user wants to search again
    search_studentid = int(input("Enter student ID: ")) #1 mark for getting user input and using it as a parameter for calling binary search function
    search_studentid_index = binary_search_recursive(students, search_studentid, 0, len(students)-1)
    if search_studentid_index == -1:
        print("Student Not Found") #1 mark for displaying "Student not found" if student not in list
    else:
        students[search_studentid_index].show_details() #1 mark for displaying student information if student in list
    search = input("Would you like to search again? Yes/No: ")
    print()
#1 mark for correct output for Student ID 15 followed by 105
```
