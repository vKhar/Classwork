```python
#Task 1.1

class Product:
    def __init__(self, s, n, p):
        self.__stockID = s
        self.__name = n
        self.__price = p

    def setStockID(self,s):
        self.__stockID = s

    def setPrice(self,p):
        self.__price = p

    def setName(self,n):
        self.__name = n

    def getStockID(self):
        return self.__stockID

    def getPrice(self):
        return self.__price

    def getName(self):
        return self.__name

```

```python
#Task 1.2

from datetime import date
class PerishableProduct(Product):
    def __init__(self, s, n, p, d):
        super().__init__(s, n, p)
        self.__date = d

    def setDate(self,d):
        self.date = d

    def getDate(self):
        return self.date

    def getPrice(self):
        expireYear,expireMonth,expireDay = self.__date.split('-') # access private

        today_date = date.today()
        new_date = date(int(expireYear),int(expireMonth),int(expireDay))
        difference = new_date - today_date
        if difference.days > 3:
            return super().getPrice() ##
        else:
            return round(super().getPrice()/100*80,2) ##
```

```python
veg = PerishableProduct("v123","kang kong", 3.00,"2023-01-01")
veg.getPrice()
```

```python
#Task 1.3
lst = []
f = open('TASK1.TXT')
for line in f:
    data = line.strip().split(',')
    if data[0] == 'P':
        lst.append(PerishableProduct(int(data[1]),data[2],float(data[3]),data[4]))
    else:
        lst.append(Product(int(data[1]),data[2],float(data[3])))
f.close()
```

```python
#task 1.4
def display(lst):
    for item in lst:
        s = ''
        s += f"{item.getStockID():<6}"
        s += f"{item.getName():<16}"
        s += f"{item.getPrice():.2f}"
        print(s)

```
