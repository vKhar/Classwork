```python
# task 2.1 Naren

def hexToDen(hex_str:str) :
  # assume that the hexacimals contains 0-9 and A-F
  total = 0
  map = "0123456789ABCDEF"
  for c in range(len(hex_str)):
     total += map.index(hex_str[c].upper()) * (16**(len(hex_str)-c-1))
  return total
```

```python
# task 2.3 part 1 Naren
class Node:
    def __init__(self, data):
        self.left_pointer = None
        self.right_pointer = None
        self.data = data

class Tree:
    def __init__(self):
        self.root = None

    def insert(self, data):
        new_node = Node(data)
        data = hexToDen(data)
        if self.root is None:
            self.root = new_node
        else:
            current = self.root
            while True:
                if data < hexToDen(current.data):
                    if current.left_pointer is None:
                        current.left_pointer = new_node
                        break
                    else:
                        current = current.left_pointer
                else:
                    if current.right_pointer is None:
                        current.right_pointer = new_node
                        break
                    else:
                        current = current.right_pointer

    def do_inorder_traversal(self, node, result_list):
        if node:
            self.do_inorder_traversal(node.left_pointer, result_list)
            result_list.append(node.data)
            self.do_inorder_traversal(node.right_pointer, result_list)

    def inorder_traversal(self):
        result_list = []
        self.do_inorder_traversal(self.root, result_list)
        return result_list
```
