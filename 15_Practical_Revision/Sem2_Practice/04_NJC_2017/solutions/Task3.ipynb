```python
#3.1
class BSTNode:
    def __init__(self, data=None):
        self.__data = data
        self.__left = None
        self.__right = None

    def get_data(self):
        return self.__data
    def set_data(self, data):
        self.__data = data 

    def get_left(self):
        return self.__left
    def set_left(self, node):
        self.__left = node 

    def get_right(self):
        return self.__right
    def set_right(self, node):
        self.__right = node 
    def __repr__(self):
        return f"{self.__data}"

class LLNode(BSTNode): ## has 3 pointers
    def __init__(self, data=None):
        super().__init__(data)
        self.__next = None
    
    def get_next(self):
        return self.__next
    def set_next(self, node):
        self.__next = node
    
```

```python
class HybridStructure:
    def __init__(self):
        self.__first = None
    def get_first(self):
        return self.__first
    def set_first(self, node):
        self.__first = node

    def inLL(self, target):
        if self.__first == None:
            return False
        cur = self.__first
        while cur :
            if (cur.get_data()//10)%10 == (target//10)%10:
                return True
            else:
                cur = cur.get_next()
        else:
            return False
            
    def insert(self, data):
        if self.inLL(data): 
            # find LLNode to insert BSTNode
            cur = self.__first
            while (cur.get_data()//10)%10 != (data//10)%10:
                cur = cur.get_next()
            ## insert into BST
            while True:
                if data < cur.get_data():
                    if cur.get_left() == None:
                        cur.set_left(BSTNode(data))
                        return
                    else:
                        cur = cur.get_left()
                else:
                    if cur.get_right() == None:
                        cur.set_right(BSTNode(data))
                        return
                    else:
                        cur = cur.get_right()                    
        else:
            ## insert into LLNode into LL
            node = LLNode(data)
            if self.__first == None:
                self.__first = node
                return            
            '''          
            # insert last
            cur = self.__first
            while cur.get_next() != None:
                cur = cur.get_next() 
            cur.set_next(node)          
            '''   
                     
            # insert in order
            cur = self.__first
            # insert before first node:
            if data < cur.get_data():
                node.set_next(cur)
                self.__first = node
            else:
            ## insert after first node
                while cur.get_next() and  cur.get_next().get_data() < data:
                    cur = cur.get_next() 
                node.set_next(cur.get_next())
                cur.set_next(node)
        
      
    def print(self):
        def in_order(node):
            if node.get_left() != None:
                in_order(node.get_left())
            print(f"{str(node):>5}")
            if node.get_right() != None:
                in_order(node.get_right())

        ## Traverse LL
        cur = self.__first
        while cur:
            # check if subtrees present
            if cur.get_left() or cur.get_right():
                in_order(cur)
                print()
            else:
                print(cur)
            cur = cur.get_next()

    def contain(self,target):
        cur_LL = self.__first
        while cur_LL:
            ## search LL 
            if target == cur_LL.get_data():
                return True
            ## check BST
            if (target//10)%10 == (cur_LL.get_data()//10)%10:
                # traverse tree
                cur_BST = cur_LL
                while cur_BST:
                    if cur_BST.get_data() == target:
                        return True
                    elif target < cur_BST.get_data():
                        cur_BST = cur_BST.get_left()
                    else:
                        cur_BST = cur_BST.get_right()
            cur_LL = cur_LL.get_next()
        else:
            return False

```
