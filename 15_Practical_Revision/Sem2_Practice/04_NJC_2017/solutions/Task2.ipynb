```python
# 2.1
import csv
def linear_search(ret, target:str)->int:
    for i in range(len(ret)):
        if ret[i][0] >= target:
            return i #insertion point
    else:
        return None ## not found

def option_1(menu_data):
    menu_data=[]
    for items in csv.reader(open("MENU.txt"),delimiter=" "):
        item_index, item_name, item_price = items[0]," ".join(items[1:-1]),items[-1]
        ## insertion sort
        insert_point = linear_search(menu_data, item_index)
        if insert_point:
            menu_data.insert(insert_point,(item_index, item_name, item_price))
        else:
            menu_data.append((item_index, item_name, item_price))
        ## 
    return menu_data
## [(item_index,item_name, item_price ),  ]
    
```

```python
menu_data = []
A = [
    ('058', "Pork Bulgogi Don", '$9.90'),
    ('059', 'Salmon Teriyaki Don', '$9.90'),
    ('040', 'Unagi Don', '$12.90'),
    ('042', 'Bibimbab Beef', '$10.90'),
    ('041', 'Bibimbab Chicken', '$9.90')
]
for item_index, item_name, item_price in A:
    insert_point = linear_search(menu_data, item_index)
    if insert_point != None:
        menu_data.insert(insert_point,(item_index, item_name, item_price))
    else:
        menu_data.append((item_index, item_name, item_price))
menu_data
```

```python
## 2.2
def bin_search(menu_data:list, lb:int, ub:int, target)-> int:
    while lb <= ub:
        mid = (lb+ub)//2
        if menu_data[mid][0] == target:
            return mid
        elif target < menu_data[mid][0]:
            ub = mid -1
        else:
            lb = mid + 1
    else:
        return -1

def option_2(menu_data):
    if menu_data == None:
        raise Exception("Menu not created")
    orders = []
    while True:
        menu_index = input("Please enter a menu item index (or -1 to quit)")
        if menu_index == "-1":
            break
        found_index  = bin_search(menu_data,0, len(menu_data)-1,menu_index)
        if found_index == -1:
            print("Invalid menu index")
        else:
            orders.append(menu_data[found_index])

    ## print orders
    ## [(item_index,item_name, item_price ),  ]
    total_price = 0
    print(f"{'Index':^8}{'Item':^10}")
    for order in orders:
        print(f"{order[0]:<8}{order[1]:<8}")
        total_price += float(order[2][1:])
    print(f"Total price: ${total_price:2.2f}")
    return menu_data ## otherwise will be replaced by None
    

def option_3(menu_data):
    raise Exception("Exit")
```

```python
## 2.1 Menu
options = ''' 
1. Read menu data
2. Take order
3. Quit
'''

# all the option function signature must be the same
actions = {
    "1":option_1,
    "2":option_2,
    "3":option_3
}
menu_data=None
while True:
    selected = input(options)
    try:
        menu_data = actions[selected](menu_data)
    except KeyError:
        print("Invalid selection")
    except  Exception as ex :
        print(f"{ex}")
        break
```
