```python
class BSTNode:
    def __init__(self,data):
        self.__data = data
        self.__left = None
        self.__right = None
        
    def get_data(self):
        return self.__data
    
    def get_left(self):
        return self.__left
    
    def get_right(self):
        return self.__right
    
    def set_data(self,data):
        self.__data = data
        
    def set_left(self,node):
        self.__left = node
        
    def set_right(self,node):
        self.__right = node
        
    def __repr__(self):
        return f'{self.__data} {self.__left} {self.__right}>'

    
    
class LLNode(BSTNode):
    def __init__(self,data):
        super().__init__(data)
        self.__next = None
    
    def get_next(self):
        return self.__next
    
    def set_next(self,node):
        self.__next = node
        
```

```python
class HybridStructure:
    def __init__(self):
        self.__first = None
    
    def get_first(self):
        return self.__first
    
    def set_first(self,node):
        self.__first = node
        
    def inLL(self,num):
        if self.__first == None:
            return False
        cur = self.__first
        while cur != None:
            curTens = str(cur.get_data())[-2]
            numTens = str(cur.get_data())[-2]
            if numTens == curTens:
                return True
            elif numTens > curTens:
                if cur.get_next() == None:
                    return False
                cur = cur.get_next()
            elif numTens < curTens:
                return False
                
    def insert(self,num):
        if self.__first == None:
            self.__first = LLNode(num)
            return True
       
        cur = self.__first
        prev = None
        while cur != None:
            curTens = str(cur.get_data())[-2]
            numTens = str(num)[-2]
            
            #if same tens got, bst insertion compare the ones
            if numTens == curTens:
                while True:
                    if num < cur.get_data():
                        if cur.get_left() == None:
                            cur.set_left(BSTNode(num))
                            return True
                        else:
                            cur = cur.get_left()
                            
                    else:
                        if cur.get_right() == None:
                            cur.set_right(BSTNode(num))
                            return True
                        else:
                            cur = cur.get_right()
                            
                        
            
            #no same tens, insert in LL
            elif numTens > curTens:
                #if nothing in next, means it is largest, insert back
                if cur.get_next() == None:
                    cur.set_next(LLNode(num))
                    return True
                #continue iterating
                prev = cur
                cur = cur.get_next()
                
            elif numTens < curTens:
                newNode = LLNode(num)
                if prev == None: #is the smallest, insert front
                    newNode.set_next(self.__first)
                    self.__first = newNode
                    return True
                else:
                #insert in the middle since smaller than the next one
                    newNode.set_next(cur)
                    prev.set_next(newNode)
                    return True
            
        
    def contains(self,num):
        cur = self.__first
        while cur != None:
            curTens = str(cur.get_data())[-2]
            numTens = str(num)[-2]   
            if curTens == numTens:
                while cur != None:
                    if cur.get_data() == num:
                        return True
                    elif cur.get_data() > num:
                        cur = cur.get_left()
                    else:
                        cur = cur.get_right()
                return False
            else:
                cur = cur.get_next()
        return False
    
    
    def __repr__(self):
        ret = ''
        LLcur = self.__first
        while LLcur != None:
            BSTcur = LLcur
            #do bst in order traversal
            def inOrder(node,l):
                if node == None:
                    return l
                inOrder(node.get_left(),l)

                l.append(node.get_data())
                inOrder(node.get_right(),l)
                return l
            l = inOrder(BSTcur,[])
            for i in range(len(l)):
                ret += str(l[i])
                ret += '\n'
            ret += '\n'
            LLcur = LLcur.get_next()
        return ret
```

```python
l = [15,17,12,13,96,91]
structure = HybridStructure()
for i in range(len(l)):
    structure.insert(l[i])
print(structure)
```
