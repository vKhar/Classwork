```python
#task 4.1

def get_hash(string,arrLen): ##no need to input arrLen, hard code 97
    total = 0
    for i in range(len(string)):
        total += ord(string[i]) * (i+1)
    return total%arrLen
```

```python
#task 4.2

loans = [None for i in range(97)]
ht1 = [None for i in range(97)]
ht2 = [None for i in range(97)]

def get_next_empty(arr,index):
    for i in range(len(arr)):
        if arr[index%len(arr)] == None:
            return index%len(arr)
        else:
            index += 1


def insert(title, borrower, loans, ht1, ht2):
    hash1 = get_hash(title,len(ht1))
    hash2 = get_hash(borrower,len(ht2))

    position = get_next_empty(loans,0)
    loans[position] = (title,borrower)
    ht1[get_next_empty(ht1,hash1)] = position
    ht2[get_next_empty(ht2,hash2)] = position
    return True               #don't return anything
```

```python

def find_loan_by_string(string,loan,pos): #pos is the index of the title/person in the tuple
    hashh = get_hash(string,len(loan))
    for i in range(len(loan)):
        if loan[hashh%len(loan)] == None: ##no need to mod hashh again
            hashh+=1
        elif loan[hashh%len(loan)][pos] == string:
            return loan[hashh%len(loan)]
        else:
            hashh += 1
    return ()

def find_loan_by_title(title,loan):
    return find_loan_by_string(title,loan,0)

def find_loan_by_person(person,loan):
    return find_loan_by_string(person,loan,1)


def delete_from_ht(hashh,index,ht):
    for i in range(len(ht)):
        if ht[hashh%len(ht)] == index:
            ht[hashh%len(ht)] = None
        else:
            hashh += 1

def remove_loan_by_string(string,loan,htString,htOther,pos):
    hashh = get_hash(string,len(loan))
    temp = hashh
    other_string = ''
    for i in range(len(loan)):
        if loan[hashh%len(loan)] == None:
            hashh += 1
        elif loan[hashh%len(loan)][pos] == string:
            other_string = loan[hashh%len(loan)][(pos+1)%2]
            loan[hashh%len(loan)] = None
            delete_from_ht(temp,hashh%len(loan),htString)
            other_hash = get_hash(other_string,len(loan))
            delete_from_ht(other_hash,hashh%len(loan),htOther)
            return True
        else:
            hashh += 1
    return False

def remove_loan_by_title(title,loan,ht1,ht2):
    return remove_loan_by_string(title,loan,ht1,ht2,0)

def remove_loan_by_person(person,loan,ht1,ht2):
    return remove_loan_by_string(person,loan,ht2,ht1,1)

```
