```python
# task 3.1
class BSTNode:
    def __init__(self, data):
        self.__data = data
        self.__left = None
        self.__right = None

    def get_data(self):
        return self.__data

    def set_data(self, data):
        self.__data = data

    def get_left(self):
        return self.__left

    def set_left(self, left):
        self.__left = left

    def get_right(self):
        return self.__right

    def set_right(self, right):
        self.__right = right

    def print(self):
        if self.__left:
            self.__left.print()
        print(self.__data)
        if self.__right:
            self.__right.print()

class LLNode(BSTNode):
    def __init__(self, data):
        super().__init__(data)
        self.__next = None

    def get_next(self):
        return self.__next

    def set_next(self, next_node):
        self.__next = next_node
```

```python
## task 3.2
class HybridStructure:
    def __init__(self):
        self.__first = None

    def get_first(self):
        return self.__first

    def set_first(self, first):
        self.__first = first

    def inLL(self, integer):
        current = self.__first
        while current:
            if (current.get_data() // 10) == (integer // 10):
                return True
            current = current.get_next()
        return False

    def insert(self, integer):
        if self.inLL(integer):
            current = self.__first
            while current:
                if (current.get_data() // 10) == (integer // 10):
                    self.insert_bst_node(current, integer)
                    return
                current = current.get_next()
        else:
            new_node = LLNode(integer)
            if self.__first is None :
                self.__first = new_node
            else :
                current = self.__first
                while current.get_next() :
                    current = current.get_next()
                current.set_next(new_node)
            #new_node.set_next(self.__first)
            #self.__first = new_node
            return

    def insert_bst_node(self, node : LLNode, data):
        if data < node.get_data():
            if node.get_left() is None:
                new_bst_node = BSTNode(data)
                node.set_left(new_bst_node)
            else:
                self.insert_bst_node(node.get_left(), data)
        else:
            if node.get_right() is None:
                new_bst_node = BSTNode(data)
                node.set_right(new_bst_node)
            else:
                self.insert_bst_node(node.get_right(), data)

    def contains(self, integer):
        current = self.__first
        while current:
            if current.get_data() == integer:
                return True
            current = current.get_next()

        return self.contains_bst_node(self.__first, integer)

    def contains_bst_node(self, node :LLNode, data):
        if node is None:
            return False
        if node.get_data() == data:
            return True
        elif data < node.get_data():
            return self.contains_bst_node(node.get_left(), data)
        else:
            return self.contains_bst_node(node.get_right(), data)

    def print(self):
        current = self.__first
        while current:
            self.print_bst_inorder(current)
            current = current.get_next()
            print("\n")

    def print_bst_inorder(self, node : LLNode):
        if node:
            self.print_bst_inorder(node.get_left())
            print(node.get_data())
            self.print_bst_inorder(node.get_right())

```
