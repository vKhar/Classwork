	    OUTPUT "Found"

		ENDIF

		IF node.left_pointer = -1 THEN

		ENDIF
		CALL Search(node.left_pointer,item)
	ENDIF

````python
hex_digits = "0123456789ABCDEF"
def gen_check_digit(digits, m):
    if len(digits) == 7:
        total = 0

        for i in range(7):
            total += int(digits[i])
        return hex_digits[(m-total%m)%m]
````

````python
## Write the validation function to check the validatitity of the input data
def validate(data, m):
    total = 0
    for i in range(8):
        total += int(data[i],16)
    if total%m == 0:
        return True
    else:
        return False
````
