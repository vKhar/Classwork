```python
#Task 1.1
#encrypt text

def encr(s): #s is a string of text
    map1 = ['A','B','C','D','E','F','G','H','I','J','K','L','M','m','a','b','c','d','e','f','g','h','i','j','k','l']
    map2 = ['N','O','P','Q','R','S','T','U','V','W','X','Y','Z','n','o','p','q','r','s','t','u','v','w','x','y','z']
    for i in range(len(s)):
        if s[i] in map1:
            return ord(s[i]) + 13
        elif s[i] in map2:
            return ord(s[i]) - 13
        else:
            return ord(s[i])


encr('A')
```
