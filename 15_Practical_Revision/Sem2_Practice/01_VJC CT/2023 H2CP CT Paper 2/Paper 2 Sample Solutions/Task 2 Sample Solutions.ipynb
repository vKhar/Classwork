```python
#Task 2.1

class Tree:
    def __init__(self, data=None, left_pointer=None, right_pointer=None):
        self.data = data
        self.left_pointer = left_pointer
        self.right_pointer = right_pointer

    def insert(self, new_data):
        if self.data is None:
            self.data = new_data
        elif new_data >= self.data:
            if self.right_pointer is None:
                self.right_pointer = Tree(new_data)
            else:
                self.right_pointer.insert(new_data)
        else:
            if self.left_pointer is None:
                self.left_pointer = Tree(new_data)
            else:
                self.left_pointer.insert(new_data)

    def print_tree(self):
        if self is None:
            return
        print(f"Data: {self.data}, Left: {self.left_pointer.data if self.left_pointer else None}, Right: {self.right_pointer.data if self.right_pointer else None}")
        if self.left_pointer:
            self.left_pointer.print_tree()
        if self.right_pointer:
            self.right_pointer.print_tree()

    def iterative_print_tree(self):
        queue = [self]
        while queue:
            current_node = queue.pop(0)
            print(f"Data: {current_node.data}, Left: {current_node.left_pointer.data if current_node.left_pointer else None}, Right: {current_node.right_pointer.data if current_node.right_pointer else None}")
            if current_node.left_pointer:
                queue.append(current_node.left_pointer)
            if current_node.right_pointer:
                queue.append(current_node.right_pointer)
```

```python
#Task 2.2

def quicksort(lst):
    def quick_sort(items, start, end):

        # Base case for recursion:
        # The recursion will stop when the partition contains a single item
        if start >= end:
            return

        # Otherwise recursively call the function
        else:
            pivot_value = items[start] # Set to first item in the partition
            low_mark = start + 1 # Set to second position in the partition
            high_mark = end # Set to last position in the partition
            finished = False

            # Repeat until low and high values have been swapped as needed
            while finished == False:

                # Move the left pivot
                while low_mark <= high_mark and items[low_mark] <= pivot_value:
                    low_mark = low_mark + 1 # Increment low_mark

                 # Move the right pivot
                while items[high_mark] >= pivot_value and high_mark >= low_mark:
                    high_mark = high_mark - 1 # Decrement high_mark

                # Check that the low mark doesn't overlap with the high mark
                if low_mark < high_mark:
                    # Swap the values at low_mark and high_mark
                    temp = items[low_mark]
                    items[low_mark] = items[high_mark]
                    items[high_mark] = temp

                # Otherwise end the loop
                else:
                    finished = True

            # Swap the pivot value and the value at high_mark
            temp = items[start]
            items[start] = items[high_mark]
            items[high_mark] = temp

            # Recursive call on the left partition
             quick_sort(items, start, high_mark - 1)

            # Recursive call on the right partition
            quick_sort(items, high_mark + 1, end)
        return items[::-1]
    return quick_sort(lst, 0, len(lst)-1)

lst = [25, 70, 20, 90, 55, 45, 65, 35, 75, 40, 30, 60, 10, 80]
sorted_lst = quicksort(lst)
print(sorted_lst)
```

```python
def quicksort(lst):
    if len(lst) <= 1:
        return lst
    else:
        pivot = lst[0]
        greater = [x for x in lst[1:] if x >= pivot]
        lesser = [x for x in lst[1:] if x < pivot]
        return quicksort(greater) + [pivot] + quicksort(lesser)

lst = [25, 70, 20, 90, 55, 45, 65, 35, 75, 40, 30, 60, 10, 80]
sorted_lst = quicksort(lst)
print(sorted_lst)
```

```python
import graphviz

class Tree:
    def __init__(self, data=None, left_pointer=None, right_pointer=None):
        self.data = data
        self.left_pointer = left_pointer
        self.right_pointer = right_pointer

    def insert(self, new_data):
        if self.data is None:
            self.data = new_data
        elif new_data >= self.data:
            if self.right_pointer is None:
                self.right_pointer = Tree(new_data)
            else:
                self.right_pointer.insert(new_data)
        else:
            if self.left_pointer is None:
                self.left_pointer = Tree(new_data)
            else:
                self.left_pointer.insert(new_data)

    def print_tree(self, graph):
        if self is None:
            return
        if self.left_pointer:
            graph.edge(str(self.data), str(self.left_pointer.data))
            self.left_pointer.print_tree(graph)
        if self.right_pointer:
            graph.edge(str(self.data), str(self.right_pointer.data))
            self.right_pointer.print_tree(graph)

def quicksort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        greater = [x for x in arr[1:] if x >= pivot]
        lesser = [x for x in arr[1:] if x < pivot]
        return quicksort(greater) + [pivot] + quicksort(lesser)

arr = [25, 70, 20, 90, 55, 45, 65, 35, 75, 40, 30, 60, 10, 80]
sorted_arr = quicksort(arr)
print(sorted_arr)

tree = Tree()
sorted_arr = quicksort([25, 70, 20, 90, 55, 45, 65, 35, 75, 40, 30, 60, 10, 80])

for val in set(sorted_arr):
    tree.insert(val)

dot = graphviz.Digraph()
tree.print_tree(dot)
dot.render('binary_search_tree', view=True)
```
