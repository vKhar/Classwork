```python
import string

upMap = string.ascii_uppercase
lowMap = string.ascii_lowercase
#print(upMap, lowMap)

def ROTchar(char):
    if char in upMap:
        index = upMap.index(char)
        return ord(upMap[(index+13)%26])
    if char in lowMap:
        index = lowMap.index(char)
        return ord(lowMap[(index+13)%26])
    else:
        return ord(char)

print(ROTchar("A"))
print(ROTchar("a"))
print(ROTchar("3"))
print(ROTchar("#"))
```

```python
userChar = input("Enter a single character") #assume user only inputs single characters
print(chr(ROTchar(userChar)))
```

```python
import string

upMap = string.ascii_uppercase
lowMap = string.ascii_lowercase
#print(upMap, lowMap)

def ROTchar(char, key):
    if char in upMap:
        index = upMap.index(char)
        return chr(ord(upMap[(index+key)%26]))
    if char in lowMap:
        index = lowMap.index(char)
        return chr(ord(lowMap[(index+key)%26]))
    else:
        return chr(ord(char))

def AROTchar(char, key):
    if char in upMap:
        index = upMap.index(char)
        return chr(ord(upMap[(index-key)%26]))
    if char in lowMap:
        index = lowMap.index(char)
        return chr(ord(lowMap[(index-key)%26]))
    else:
        return chr(ord(char))


```

```python
x = "A"
enc_x = ROTchar(x, 999)
print(AROTchar(enc_x, 999))
```
