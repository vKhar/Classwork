```python
#Task 2.1
class Tree:
    def __init__(self):
        self.data = None
        self.left_pointer = None
        self.right_pointer = None
    def insert(self,data):
        if self.data == None:
            self.data = data
            return
        if data > self.data:
            if self.right_pointer == None:
                new_node = Tree()
                new_node.data = data
                self.right_pointer = new_node
            else:
                self.right_pointer.insert(data)
        else:
            if self.left_pointer == None:
                new_node = Tree()
                new_node.data = data
                self.left_pointer = new_node
            else:
                self.left_pointer.insert(data)

    def print_tree(self):
        print(self.data)
        if self.left_pointer != None:
            self.left_pointer.print_tree()
        if self.right_pointer != None:
            self.right_pointer.print_tree()

```

```python
#Task 2.2
def quicksort(lst):
    if len(lst) < 2:
        return lst
 
    ## random pivot 
    import random
    pivot_i = random.randint(0, len(lst)-1)
    lst[0], lst[pivot_i] = lst[pivot_i],lst[0]
    pivot = lst.pop(0)
    left = []
    right = []
    for digit in lst:
        if digit < pivot:
            left.append(digit)
        else:
            right.append(digit)
    return quicksort(right) + [pivot] + quicksort(left)

print(quicksort([25, 70, 20, 90, 55, 45, 65, 35, 75, 40, 30, 60, 10, 80]))
```
