```python
import csv
from datetime import datetime, timedelta
def task1_1(filename):
    datlist=[]
    with open(filename) as f:
        datlist=list(csv.reader(f))[1:]
    for i in datlist:
        dates=i[0].split('-')
        i.append(datetime(int(dates[0]), int(dates[1]), 1))
    for pointer in range(1,len(datlist)):
        while pointer>0 and datlist[pointer][2]<datlist[pointer-1][2]:
            datlist[pointer], datlist[pointer-1]=datlist[pointer-1], datlist[pointer]
            pointer-=1
    ret=[]
    for i in datlist:
        if "2023" in i[0]:
            pass
        else:
            ret.append(i[0:2])
    return ret
```

```python
print(task1_1("SUNSHINE.csv"))
```

```python
def task1_2(sortedlist):
    yearDic={}
    for i in sortedlist:
        year=i[0][0:4]
        if year not in yearDic:
            yearDic[year]=[[i[0][5:],float(i[1])]]
        else:
            yearDic[year].append([i[0][5:],float(i[1])])
    for x in list(yearDic.keys()):
        low=[0, 999999]
        high=[0, -1]
        for i in yearDic[x]:
            if i[1]>high[1]:
                high=i
            if i[1]<low[1]:
                low=i
        print(x)
        print(tuple(low))
        print(tuple(high))
        print(high[1]-low[1])
task1_2(task1_1("SUNSHINE.csv"))
```

```python
from statistics import mean
def task1_3(sortedlist):
    sumL={"01":[], '02':[], '03':[], '04':[], '05':[], '06':[], '07':[], '08':[], '09':[], '10':[], "11":[], '12':[]}
    for i in sortedlist:
        sumL[i[0][5:]].append(float(i[1]))
    for i in sumL.keys():
        print(f"Month {i}: average daily sunshine duration 1982-2022 was {mean(sumL[i])} hours.")
task1_3(task1_1("SUNSHINE.csv"))
```
