```python
ten_cent_coins = int(input("Enter the number of 10-cent coins inserted:"))
twenty_cent_coins = int(input("Enter the number of 20-cent coins inserted:"))
fifty_cent_coins = int(input("Enter the number of 50-cent coins inserted:"))
one_dollar_coins = int(input("Enter the number of 1-dollar coins inserted:"))
price = float(input("Please enter either price of the drink, i.e, 0.8 or 1.2:"))
total_inserted = (ten_cent_coins*0.1+twenty_cent_coins*0.2+fifty_cent_coins*0.5+one_dollar_coins)
print(f"Total inserted: ${total_inserted}")
##Checks if the amount inserted is enough
change = float(total_inserted-price)
if change < 0 :
    print("You do not have enough money to pay for the drink!")
else :
    print(f"The machine returns a total of ${change}, in the form of:")
##These variables represent the amount of coins returned
tens_returned = 0
twenties_returned = 0
fifties_returned = 0
ones_returned = 0
while change >= 1 :
    change -= 1
    ones_returned += 1
    
while change >= 0.5 :
    change -= 0.5
    fifties_returned += 1
    
while change >= 0.2 :
    change -= 0.2
    twenties_returned += 1
    
while change > 0 :
    change -= 0.1
    tens_returned += 1
print(f"{ones_returned} x 1-dollar coin(s)")
print(f"{fifties_returned} x 50-cent coin(s)")
print(f"{twenties_returned} x 20-cent coin(s)")
print(f"{tens_returned} x 10-cent coin(s)")
print("Thank you for your purchase!")
```

### Question 3B: Unit Converter
By Gary & Ta Wenn

```python
#setup dictionary to point to relative values
baseUnitToOperationMap = {"mm": 1, "cm": 10, "m": 1000, "km": 1000000}
typeToOperationMap = {"length": 1, "area": 2, "volume": 3}
#catch any errors if input cannot be casted
try:
    #gather input
    quantity = float(input("Input a quantity"))
    typeOfQuantity = input("Select either length, area or volume")
    initialBaseUnit = input("Select the base unit of measurement (mm, cm , m, km)")
    desiredBaseUnit = input("Select the desired unit of measurement (mm, cm , m, km)")
    #if type of quantity, initial base unit and desired base unit not valid, don't run
    if typeOfQuantity not in ['length', 'area', 'volume'] or initialBaseUnit not in ['mm', 'cm' ,'m', 'km'] or desiredBaseUnit not in ['mm', 'cm' ,'m', 'km'] or quantity < 0:
        print("Please input valid values")
    else:
        #calculate result
        result = quantity * ((baseUnitToOperationMap[initialBaseUnit] / baseUnitToOperationMap[desiredBaseUnit])**typeToOperationMap[typeOfQuantity])
        print(f"{quantity} {initialBaseUnit}^{typeToOperationMap[typeOfQuantity]} = {round(result, 6)} {desiredBaseUnit}^{typeToOperationMap[typeOfQuantity]}")
except:
    print("Please input valid values")
```
