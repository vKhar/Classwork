```python
import socket
listen_socket = socket.socket()
listen_socket.bind(('', 80))
listen_socket.listen()
print("server waiting ...")
s, addr = listen_socket.accept()
data = s.recv(1000000)
print(data.decode())
s.sendall(b"<html><h1>Bye</h1></html>")
print("server connection closed")
s.close()
listen_socket.close()
```
